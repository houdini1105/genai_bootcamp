```python
# Challenge 1: Filtering Words
# Write a Python program that iterates through the following list of 
# fruits: ['apple', 'banana', 'grape', 'kiwi', 'orange']. Use a for loop and 
# the continue statement to print only the fruits that do not contain 
# the letter 'a'.
fruits = ['apple', 'banana', 'grape', 'kiwi', 'orange']
for char in fruits:
    if 'a' not in char:
        print(char)
# Using the continue as challenged
fruits = ['apple', 'banana', 'grape', 'kiwi', 'orange']
for char in fruits:
    if 'a' in char:
        continue
    print(char)

# Challenge 2: Number Skipping
# Create a program that uses a for loop to iterate through numbers from 1 to 10. 
# If the number is even, use continue to skip printing it. Print all the odd 
# numbers in the range.
for n in range(1, 11):
    if n % 2 == 0:
        continue
    else:
        print(n)

# Challenge 3: Simple Countdown
# Write a while loop that counts down from 5 to 1, printing each number. 
# After the loop finishes, print "Blast Off!".
count = 5
while count >= 1:
    print(count)
    count -= 1
print('Blast Off!')

# Challenge 4: Password Prompt
# Design a program that repeatedly asks the user to enter a password. The 
# loop should continue until the user enters "secret123". Once the correct 
# password is entered, print "Access granted!" and exit the loop using break.
while True:
    user_prompt = input('Enter a password:')
    if user_prompt == 'secret123':
        print('Access Granted')
        break

# Challenge 5: Input Length Check
# Create a while loop that continuously prompts the user to enter a sentence. 
# If the sentence has fewer than 10 characters, print "Sentence too short. 
# Please try again." and use continue to ask for input again. If the 
# sentence is 10 or more characters, print "Sentence accepted!" and break 
# out of the loop.
while True:
    sent = input('Enter a sentence:')
    if len(sent) <= 9:
        print('"Sentence too short. Please try again.')
        continue
    print('Sentence accepted!')
    break

# Challenge 6: Placeholder Function
# Define a function called process_data that takes a list as an argument. 
# Inside the function, use a for loop to iterate through the list, but for 
# now, just use the pass statement as a placeholder. Print 
# "Function process_data called." after the loop (but still inside the function).
l1 = []
def process_data(l1):
    for x in l1:
        pass
    print('Function process-data called.')

# Challenge 7: Even Number Sum (with continue)
# Write a program that calculates the sum of all even numbers 
# from 1 to 20 (inclusive). Use a for loop and the continue statement to 
# skip odd numbers.
sum = 0
for n in range(1, 21):
    if n % 2 != 0:
        continue
    if n % 2 == 0:
        sum += n
print(sum)

# Challenge 8: Temperature Monitor (with break and else)
# Simulate a temperature monitor. Start with current_temp = 65. Use a 
# while loop to increase the current_temp by 1 in each iteration and print 
# the temperature. If the current_temp reaches 72, print "Temperature threshold 
# reached. Cooling down." and break the loop. If the loop completes 
# without breaking (i.e., it reaches its natural end, perhaps current_temp 
# goes above 75), print "Temperature monitoring complete."
current_temp = 65
while current_temp <= 75:
    print(current_temp)
    current_temp += 1
    if current_temp == 72:
        print('Temperature threshold reached. Cooling down.')
        break
else:
    print('Tempature monitoring complete.')

# Challenge 9: Search for a Keyword
# You have a list of articles: ['AI in Healthcare', 'The Future of Robotics', 'Machine Learning Basics', 'AI Revolution in Business']. 
# Iterate through this list using a for loop. If you find an article title 
# that contains "Revolution", print "Found a key article!" and break the loop. 
# Otherwise, just print the article title.
articles = ['AI in Healthcare', 'The Future of Robotics', 'Machine Learning Basics', 'AI Revolution in Business']
for char in articles:
    if 'Revolution' in char:
        print('Found a key article!')
        break
    else:
        print(char)

# Challenge 10: Nested Loop with break
# Create a nested for loop. The outer loop should iterate from i = 0 to 2. 
# The inner loop should iterate from j = 0 to 2. If j is equal to 1, use break 
# in the inner loop. Print the values of i and j in each iteration before 
# the inner loop's break condition.
for i in range(0, 3):
    for j in range(0, 3):
        print(i, j)
        if j == 1:
            break

# Challenge 11: Infinite Loop Prevention
# Write a while loop that asks the user to enter a positive number. If the 
# user enters a non-positive number (zero or negative), print an error message 
# and use continue to prompt again. If a positive number is entered, 
# print "Valid number entered!" and break the loop.

while True:
    user_num = int(input('Enter a positive num:'))
    if user_num <= 0:
        print('Error! Enter a positive #:')
        continue
    else:
        print('Positive # entered')
        break

# Challenge 12: List Processing with pass
# You are given a list of data: data = [10, 25, 5, 30, 15]. Write a for loop 
# to iterate through this list. For now, inside the loop, just use the 
# pass statement. After the loop, print "Data processing setup complete."
data = [10, 25, 5, 30, 15]
for num in data:
    pass
print('Data processing setup complete.')

# Challenge 13: Character Counter (with continue)
# Ask the user to enter a string. Iterate through each character of the 
# string using a for loop. If the character is a space, use continue to 
# skip counting it. Count and print the total number of non-space characters 
# in the string.
user_string = input('Enter a sent:')
count = 0
for i in user_string:
    if i == ' ':
        continue
    else:
        count += 1
print(count)

# Challenge 14: Login Attempts
# Simulate a login system. Allow the user a maximum of 3 login attempts. Use a 
# while loop. In each attempt, ask for a username and 
# password (e.g., "admin" and "password123"). If both are correct, 
# print "Login successful!" and break. If incorrect, 
# print "Incorrect credentials. Attempts remaining: X" (where X decreases). If 
# 3 attempts are used without success, print "Too many failed attempts. 
# Account locked."
login_attempts = 3
user_id = 'admin'
pw = 'password123'
while login_attempts > 0:
    user = input('Enter userid:')
    user_pw = input('Enter password:')
    if user == user_id and user_pw == pw:
        print('Log-in successful')
        break
    else:
        login_attempts -= 1
        print(f'Incorrect credentials. Attempts remaining: {login_attempts}')
else:
    print('Too many failed attempts. Account locked.')

# Challenge 15: Nested Loop with Conditional continue
# Create a nested for loop where the outer loop iterates from row = 1 to 3 and 
# the inner loop iterates from col = 1 to 3. If row is 2 and col is 2, 
# use continue to skip printing that specific combination. For all other 
# combinations, print (row, col).
for row in range(1, 4):
    for col in range(1, 4):
        if row == 2 and col == 2:
            continue
        print(row, col)
```
