```python
# Challenge 1: Counting Up to 10
# Write a Python program that uses a while loop to count from 
# 1 up to 10, printing each number.
num = 1
while num <= 10:
    print(num)
    num += 1

# Challenge 2: Sum of Numbers
# Create a program that uses a while loop to calculate the sum of 
# integers from 1 to 5 (i.e., 1 + 2 + 3 + 4 + 5). Print the final sum.
current_num = 1
running_total = 0
while current_num <= 5:
    running_total += current_num
    current_num += 1
print(running_total)

# Challenge 3: Guess the Secret Number
# Write a program that sets a secret_number (e.g., 7). Use a while 
# loop to repeatedly ask the user to "Guess the number (between 1 and 10):". 
# If the user guesses correctly, print "Congratulations! You guessed it!" 
# and use break to exit the loop.

secret_num = 7

while True:
    user_num = int(input('Guess the num between 1 and 10:'))
    if user_num == secret_num:
        print('Congragulations! You guessed it.')
        break

# Challenge 4: Positive Input Only
# Create a program that uses a while loop to continuously prompt the 
# user to "Enter a positive number:". If the user enters a number less 
# than or equal to zero, print "Invalid input. Please try again." and use 
# continue to ask for input again. Once a positive number is entered, 
# print "Thank you for the positive number!" and exit the loop.
while True:
    user_num = int(input('Enter a positive #:'))
    if user_num <= 0:
        print('Invalid input. Please try again.')
        continue
    else:
        print('Thank you for the positive number.')
        break

# Challenge 5: Even Number Generator
# Write a while loop that starts with a number (e.g., start = 2) and prints 
# the next five even numbers.
start = 2
count = 0
while True:
    print(start)
    start = start + 2
    count += 1
    if count == 5:
        break

# Challenge 6: Repeat a Character
# Ask the user to enter a character and a number. Use a while loop to print 
# that character the specified number of times on a single line 
# (without spaces between characters).
char = input('Enter a char:')
num = int(input('Enter a num:'))
count = 0
while True:
    print(char, end='')
    count += 1
    if count == num:
        break

# Challenge 7: Countdown to Zero
# Write a while loop that asks the user to "Enter a starting number for 
# the countdown:". Then, count down from that number to 0, printing each number. 
# After 0, print "Liftoff!".
start = int(input('Enter a num:'))
while start >= 0:
    print(start)
    start -= 1
print('Liftoff!')

# Challenge 8: Find First Divisible
# Start with a number num = 10. Use a while loop to find the first 
# number greater than or equal to num that is divisible by 7. Print that number.
num = 10
while num % 7 != 0:
    num += 1
print(num)

# Challenge 9: Simple Menu Interaction
# Create a program that displays a simple menu:

# Menu:
# 1. Play Game
# 2. View High Scores
# 3. Exit
# Enter your choice:
# Use a while loop to keep displaying the menu and asking for input 
# until the user enters '3' to exit. For choices '1' and '2', just 
# print "Playing game..." or "Viewing high scores..." respectively.
while True:
    print('Menu:\n 1. Play Game\n 2. View High Scores\n 3. Exit')
    user_input = int(input('Enter your choice:'))
    if user_input == 1:
        print('Play Game')
    elif user_input == 2:
        print('View High Score')
    elif user_input == 3:
        print('Exit')
        break
    else:
        print('Invalid selection. Select 1, 2 or 3.')

# Challenge 10: Iterative Factorial
# Write a program that asks the user to enter a non-negative integer. 
# Use a while loop to calculate the factorial of that 
# number (e.g., factorial of 5 is 5*4*3*2*1 = 120). Print the 
# result. (Assume valid non-negative integer input for simplicity).
user_num = int(input('Enter a num:'))
result = 1
while user_num >= 1:
    result *= user_num
    user_num -= 1
print(result)
```
