Challenge 1: Extract Alphabets Only
Problem: Write a Python function extract_alphabets(text) that takes a string text and returns a new string containing only the alphabetic characters from the input. All characters in the returned string should be lowercase.

Example:
extract_alphabets("Hello World! 123") should return "helloworld".
extract_alphabets("Pyth0n Pr0gramm!ng") should return "pythonprogramming".

Challenge 2: Remove Digits from String
Problem: Write a Python function remove_digits(text) that takes a string text and returns a new string with all digit characters removed.

Example:
remove_digits("Product ID: P12345") should return "Product ID: P".
remove_digits("2024-07-08") should return "--".

Challenge 3: Count Vowels and Consonants
Problem: Write a Python function count_vowels_consonants(text) that takes a string text. It should return a tuple (vowel_count, consonant_count) representing the total number of vowels (a, e, i, o, u, case-insensitive) and consonants in the string. Ignore any non-alphabetic characters.

Example:
count_vowels_consonants("Hello World") should return (4, 6).
count_vowels_consonants("Python 3.9") should return (1, 5).

Challenge 4: First Vowel Index
Problem: Write a Python function find_first_vowel(word) that takes a string word and returns the index of the first occurrence of any lowercase vowel ('a', 'e', 'i', 'o', 'u'). If no vowel is found, return -1.

Example:
find_first_vowel("rhythm") should return -1.
find_first_vowel("beautiful") should return 2.

Challenge 5: Contains All Vowels
Problem: Write a Python function has_all_vowels(text) that returns True if the given text contains at least one instance of each of the five lowercase vowels ('a', 'e', 'i', 'o', 'u'), and False otherwise. Case should be ignored when checking for vowels.

Example:
has_all_vowels("education") should return True.
has_all_vowels("hello world") should return False (missing 'a', 'i', 'u').

Challenge 6: Find Second Occurrence of Character
Problem: Write a Python function find_second_occurrence(text, char) that takes a string text and a single character char. It should return the index of the second occurrence of char in text. If char appears less than two times, return -1.

Example:
find_second_occurrence("banana", "a") should return 3.
find_second_occurrence("programming", "m") should return 6.
find_second_occurrence("hello", "l") should return 3.
find_second_occurrence("apple", "p") should return -1.

Challenge 7: Check for Consecutive Duplicates
Problem: Write a Python function has_consecutive_duplicates(text) that returns True if the string text contains any two identical characters next to each other, and False otherwise.

Example:
has_consecutive_duplicates("hello") should return True (because of "ll").
has_consecutive_duplicates("programming") should return True (because of "mm").
has_consecutive_duplicates("python") should return False.

Challenge 8: Censor Specific Word
Problem: Write a Python function censor_word(sentence, word_to_censor) that takes a sentence string and a word_to_censor string. It should replace all occurrences of word_to_censor in the sentence with asterisks (*), maintaining the original length of the censored word. The comparison should be case-insensitive.

Example:
censor_word("This is a secret message.", "secret") should return "This is a ****** message.".
censor_word("Bad words are bad.", "bad") should return "*** words are ***.".

Challenge 9: Replace All Occurrences (Case Insensitive)
Problem: Write a Python function replace_case_insensitive(text, old_sub, new_sub) that replaces all occurrences of old_sub with new_sub in text, ignoring the case of old_sub. The case of new_sub should be preserved.

Example:
replace_case_insensitive("Hello World, hello Python", "hello", "Hi") should return "Hi World, Hi Python".
replace_case_insensitive("APPLE pie, apple sauce", "apple", "ORANGE") should return "ORANGE pie, ORANGE sauce".

Challenge 10: Replace First N Occurrences
Problem: Write a Python function replace_first_n(text, old_sub, new_sub, n) that replaces only the first n occurrences of old_sub with new_sub in text. If old_sub appears fewer than n times, replace all occurrences.

Example:
replace_first_n("one two one three one", "one", "ZERO", 2) should return "ZERO two ZERO three one".
replace_first_n("a a a a a", "a", "X", 3) should return "X X X a a".

Challenge 11: Extract Middle Characters
Problem: Write a Python function get_middle_chars(s) that takes a string s.

If the length of s is odd, return the single middle character.

If the length of s is even, return the two middle characters.

Example:
get_middle_chars("apple") should return "p".
get_middle_chars("python") should return "th".
get_middle_chars("a") should return "a".

Challenge 12: Swap First and Last Characters
Problem: Write a Python function swap_first_last(s) that takes a string s and returns a new string where the first and last characters are swapped. If the string has less than two characters, return the original string.

Example:
swap_first_last("hello") should return "oellh".
swap_first_last("a") should return "a".
swap_first_last("ab") should return "ba".

Challenge 13: Remove Middle Section by Indices
Problem: Write a Python function remove_section(s, start_index, end_index) that takes a string s and two integer indices. It should return a new string with the characters from start_index (inclusive) up to end_index (exclusive) removed. Assume valid indices within the string's bounds.

Example:
remove_section("programming", 3, 7) should return "praming" (removes "gram").
remove_section("abcdefg", 0, 2) should return "cdefg".

Challenge 14: Extract Every Nth Character
Problem: Write a Python function every_nth_char(text, n) that returns a new string containing every n-th character from text, starting with the first character (index 0). If n is 0 or less, return an empty string.

Example:
every_nth_char("abcdefghijkl", 3) should return "adgj".
every_nth_char("programming", 2) should return "pormn".
every_nth_char("hello", 1) should return "hello".

Challenge 15: Palindrome Check (Advanced Slicing)
Problem: Write a Python function is_palindrome_slicing(text) that checks if a given text is a palindrome (reads the same forwards and backwards). The function should ignore case and non-alphabetic characters, and must use string slicing to reverse the cleaned string.

Example:
is_palindrome_slicing("Racecar") should return True.
is_palindrome_slicing("A man, a plan, a canal: Panama") should return True.
is_palindrome_slicing("hello world") should return False.

Challenge 16: Sanitize Input String
Problem: Write a Python function sanitize_input(raw_input) that takes a raw string input. It should perform the following cleaning steps in order:

Remove leading and trailing whitespace.

Convert the entire string to lowercase.

Remove all non-alphanumeric characters (keep only letters and digits).

Example:
sanitize_input("  Hello World! 123@ ") should return "helloworld123".
sanitize_input("   Python-Programming_v1.0 ") should return "pythonprogrammingv10".

Challenge 17: Simple Caesar Cipher Encoder
Problem: Write a Python function caesar_encode(text, shift) that implements a simple Caesar cipher. It should shift each alphabetic character in text forward by shift positions in the alphabet. Non-alphabetic characters should remain unchanged. The shift should wrap around the alphabet (e.g., 'z' shifted by 1 becomes 'a'). Maintain original case.

Example:
caesar_encode("abc", 1) should return "bcd".
caesar_encode("xyz", 3) should return "abc".
caesar_encode("Hello, World!", 3) should return "Khoor, Zruog!".

Challenge 18: Is Valid Password?
Problem: Write a Python function is_valid_password(password) that checks if a given password string meets the following criteria:

Minimum length of 8 characters.

Contains at least one uppercase letter.

Contains at least one lowercase letter.

Contains at least one digit.

Return True if all criteria are met, False otherwise.

Example:
is_valid_password("Password123") should return True.
is_valid_password("password") should return False (no uppercase, no digit).
is_valid_password("P@ss1") should return False (too short).

Challenge 19: Format Phone Number (Advanced)

It should extract only the digits and then format them into a standard "XXX-XXX-XXXX" format. Assume the input will always contain exactly 10 digits.

Example:
format_phone_number_advanced("123 456 7890") should return "123-456-7890".
format_phone_number_advanced("(987)654-3210") should return "987-654-3210".
format_phone_number_advanced("987.654.3210") should return "987-654-3210".
