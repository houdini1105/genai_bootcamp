```python
# Challenge 1: Simple Counter
# Goal: Write a function print_numbers_up_to(n) that takes a positive integer n 
# and prints every number from 1 up to n (inclusive).
def print_numbers_up_to(n):
    current_number = 1  # Initialized counter: Starts at 1
    while current_number <= n: # While loop condition: Continues as long as current_number is within range
        print(current_number)   # Action: Print the current number
        current_number += 1     # Increment counter: Moves to the next number, ensures loop ends

# Challenge 2: Summing Numbers
# Goal: Write a function sum_up_to(n) that takes a positive integer n and returns 
# the sum of all integers from 1 up to n. (e.g., if n=3, it should return 1+2+3=6).
def sum_up_to(n):
    current_number = 1
    total_sum = 0
    while current_number <= n:
        total_sum += current_number
        current_number += 1
    return total_sum

# Challenge 3: First Even Number
# Goal: Write a function find_first_even(start_num) that takes an integer start_num and 
# finds the first even number that is greater than or equal to start_num. It 
# should return this first even number. If start_num is extremely large and no 
# even number is found within a reasonable range (let's say up to start_num + 100 to 
# prevent infinite loops), return None.
def find_first_even(start_num):
    current_check = start_num

    # Step 1: Handle the case where start_num is already even
    if current_check % 2 == 0:
        return current_check

    # Step 2: If start_num was odd, increment current_check until it becomes even
    # or exceeds the limit.
    while current_check % 2 != 0 and current_check <= start_num + 100:
        current_check += 1

    # Step 3: After the loop, determine WHY it stopped and return accordingly.
    # The loop stopped for one of two reasons:
    # A) current_check became even (current_check % 2 != 0 became False)
    # B) current_check exceeded the limit (current_check <= start_num + 100 became False)
    if current_check % 2 == 0:
        # If it's even, it means we found one (and the loop stopped because of it)
        return current_check
    else:
        # If it's not even, it means the loop stopped because it hit the limit
        # (and remained odd, or it was odd when it first exceeded the limit)
        return None
```

```python
# Mini-Challenge 1: Counting Up
# Goal: Write a function count_up(limit) that prints numbers starting 
# from 1 up to the limit (inclusive).

def count_up(limit):
    current_num = 1
    while current_num <= limit:
        print(current_num)
        current_num += 1

# Mini-Challenge 2: Counting Down
# Goal: Write a function count_down(start_num) that prints numbers starting from 
# start_num down to 1 (inclusive).
def count_down(start_num):
    current_num = start_num
    while current_num >= 1:
        print(current_num)
        current_num -= 1

# Mini-Challenge 3: Multiples of 5
# Goal: Write a function print_multiples_of_5(limit) that prints multiples of 5, 
# starting from 5, up to the limit (inclusive).
def print_multiples_of_5(limit):
    start_num = 5                # Initialized counter to starting multiple
    while start_num <= limit:    # While loop condition (within limit)
        print(start_num)         # Action inside loop (print)
        start_num += 5           # Increment counter by 5 (to get next multiple)

# New Challenge 1: Print Evens in Range
# Goal: Write a function print_evens_in_range(start, end) that prints all even numbers 
# starting from start up to end (inclusive).
def print_evens_in_range(start, end):
    current_number = start
    while current_number <= end:
        # Step 1 & 2: Check the current_number and print if it's even
        if current_number % 2 == 0:
            print(current_number)

        # Step 3: ALWAYS move to the next number for the next loop iteration
        # This line must be outside the 'if' block but still inside the 'while' loop
        current_number += 1
            
```

```python
# New Challenge 2: Character Repeater
# Goal: Write a function repeat_char(char, times) that takes a single character 
# char (a string of length 1) and an integer times, and prints that char exactly 
# times number of times, each on a new line.
def repeat_char(char, times):
    counter = 0
    while times > counter:
        print(char)
        counter += 1
```
