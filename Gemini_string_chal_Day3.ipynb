```python
# Part 1: String Fundamentals & Basic Properties
# Coding Challenge: Get User's Name Length

# Task: Write a Python script that asks the user for their first name and then prints the total number of characters in the name.
name = input('Enter your name:')
len_name = len(name)
print(f'Your name {name} is {len_name} characters')
# Example Output:
# What is your first name? Alice
# Your name has 5 characters.

# Coding Challenge: First and Last Letter
# Task: Write a Python function get_first_last_char(word) that takes a string word and returns a new string containing only its first and last characters. Assume the input word will always have at least one character.
# Example: get_first_last_char("Python") should return "Pn".
# Get input from the user (this part is outside the function)
word = input('Enter a word:')

# Define the function
def get_first_last_char(word):
    """
    This function takes a word and returns a new string
    containing only its first and last characters.
    """
    first_char = word[0]
    last_char = word[-1]
    new_word = first_char + last_char
    return new_word # No need for parentheses around new_word here, though it works.

# Call the function with the user's input and store the result
result = get_first_last_char(word)

# Print the result returned by the function
print(result)
# The below method works however the challange is to create a function
# word = input('Enter a word:')
# first_char = word[0]
# last_char = word[-1]
# new_word = first_char + last_char
# print(new_word)

# Coding Challenge: Greeting Creator
# Task: Write a Python function create_greeting(name) that takes a string name and returns a greeting like "Hello, [Name]!".
# Example: create_greeting("Alice") should return "Hello, Alice!".
# Get input from the user
name = input('Enter your name:')

# Define the function
def create_greeting(name):
    """
    This function takes a name and returns a greeting like "Hello, [Name]!".
    """
    # Correctly includes the comma and exclamation mark as per example
    greeting = 'Hello, ' + name.capitalize() + '!'
    return greeting

# Call the function and print the result
result = create_greeting(name)
print(result)

# Coding Challenge: Is String Empty?
# Task: Write a Python function is_empty(text) that returns True if the given 
# string text is completely empty (has no characters), and False otherwise.
# Example: is_empty("") should return True. is_empty(" ") should return False.
text = input('Enter a word:')

# Define the function
def is_empty(text):
    """
    This function returns True if the given string text is completely empty (has no characters),
    and False otherwise.
    """
    if len(text) == 0:
        return True # No need for parentheses around True/False, but it works
    else:
        return False

# --- Testing the function ---
# Get input from the user (optional, you can also just test directly)
user_input = input('Enter a word (or nothing):')
is_it_empty = is_empty(user_input)
print(f"Is '{user_input}' empty? {is_it_empty}")

# Test with examples from the challenge:
print(f"Is '' empty? {is_empty('')}")       # Expected: True
print(f"Is ' ' empty? {is_empty(' ')}")     # Expected: False
print(f"Is 'hello' empty? {is_empty('hello')}") # Expected: False

# A more Pythonic way
def is_empty_pythonic(text):
    """
    This function returns True if the given string text is completely empty (has no characters),
    and False otherwise, in a more concise way.
    """
    return len(text) == 0 # This directly returns True or False

# Coding Challenge: Title Case Converter
# Task: Write a Python function make_title_case(sentence) that takes a string 
# sentence and returns a new string where the first letter of each word 
# is capitalized and the rest are lowercase.
# Example: make_title_case("hello world") should return "Hello World".

# Get input from the user
sentence = input('Enter a sentence: ')

# Define the function
def make_title_case(sentence):
    """
    This function takes a string and returns a new string
    where the first letter of each word is capitalized.
    """
    formatted_sentence = sentence.title()
    return formatted_sentence

# Call the function with the user's input and print the result
result = make_title_case(sentence)
print(result)

# Coding Challenge: Clean Up Spaces
# Get input from the user
sentence = input('Enter a sent:')

# Define the function
def trim_string(sentence):
    """
    This function takes a string and returns a new string
    with any leading or trailing whitespace removed.
    """
    clean_string = sentence.strip()
    return clean_string

# Call the function and print the result
result = trim_string(sentence)
print(result)

# Part 2: Substrings & Core String Methods
# Coding Challenge: Contains "cat"?
# Task: Write a Python function has_cat(animal_string) that returns True if 
# the string animal_string contains the substring "cat" (case-sensitive), 
# and False otherwise.
# Example: has_cat("My cat is cute") should return True. has_cat("My dog is cute") should return False.
animal_string = input('Enter a string').lower()

def has_cat(animal_string):
    if 'cat' in animal_string:
        return True
    else:
        return False
    
result = has_cat(animal_string)
print(result)

def has_cat_pythonic(animal_string):
    return 'cat' in animal_string # This directly returns True or False

# Coding Challenge: Find 'o' Position
# Task: Write a Python function find_o_position(word) that takes a string word. 
# It should return the index of the first occurrence of the letter 'o' (lowercase). 
# If 'o' is not found, return -1.
# Example: find_o_position("hello") should return 4. 
# find_o_position("Python") should return -1.
# Get input from the user (for testing)
word = input('Enter a word: ')

# Define the function
def find_o_position(word):
    """
    This function returns the index of the first occurrence of 'o' (lowercase).
    Returns -1 if 'o' is not found.
    """
    # Get the result of find() first, as we'll use it twice
    found_index = word.find('o')

    if found_index != -1:
        return found_index # Return the actual index
    else:
        return -1 # Return -1 if not found

# Call the function and print the result
result = find_o_position(word)
print(f"The first 'o' is at index: {result}")

# Test cases from the example:
print(f"Position of 'o' in 'hello': {find_o_position('hello')}") # Expected: 4
print(f"Position of 'o' in 'Python': {find_o_position('Python')}") # Expected: -1
print(f"Position of 'o' in 'orange': {find_o_position('orange')}") # Expected: 0


# Coding Challenge: Replace Vowels with 'X'
# Task: Write a Python function replace_vowels_with_x(text) 
# that takes a string text and replaces all lowercase 
# vowels (a, e, i, o, u) with the letter 'X'.
# Example: replace_vowels_with_x("apple") should return "XpplX".

# Get input from the user
text = input('Enter a string: ')

# Define the function
def replace_vowels_with_x(text):
    """
    This function replaces all lowercase vowels (a, e, i, o, u)
    in a string with the letter 'X'.
    """
    # Perform replacements for each vowel sequentially
    new_text = text.replace('a','X')
    new_text = new_text.replace('e','X')
    new_text = new_text.replace('i','X')
    new_text = new_text.replace('o','X')
    new_text = new_text.replace('u','X')
    return new_text

# Call the function and print the result
result = replace_vowels_with_x(text)
print(result)

# Coding Challenge: Count 'a's
# Task: Write a Python function count_a_letters(sentence) that takes a 
# string sentence and returns the number of times the letter 'a' (lowercase) 
# appears in it.
# Example: count_a_letters("banana") should return 3.
# Get input from the user
sentence = input('Enter a sentence: ')

# Define the function
def count_a_letter(sentence):
    """
    This function counts the number of times the lowercase letter 'a'
    appears in the given sentence.
    """
    return sentence.count('a')

# Call the function and print the result
result = count_a_letter(sentence)
print(f"The letter 'a' appears {result} times.")

# Coding Challenge: Does it Start with "Mr."?
# Task: Write a Python function is_mr(name) that returns True if a 
# given name string starts with "Mr." (case-sensitive), and False otherwise.
# Example: is_mr("Mr. Smith") should return True.
# Get input from the user
name = input('Enter a name: ')

# Define the function
def is_mr(name):
    """
    This function returns True if a name string starts with "Mr." (case-sensitive),
    and False otherwise.
    """
    # The startswith() method directly returns True or False
    new_name = name.startswith('Mr.')
    return new_name

# Call the function and print the result
result = is_mr(name)
print(result)name = input('Enter a name:')

def is_mr(name):
    new_name = name.startswith('Mr.')
    return new_name

result = is_mr(name)
print(result)

# Coding Challenge: Does it End with "ing"?
# Task: Write a Python function ends_with_ing(verb) that returns 
# True if a given verb string ends with "ing" (case-sensitive), 
# and False otherwise.
# Example: ends_with_ing("running") should return True.

# Get input from the user
verb = input('Enter a verb: ')

# Define the function
def ends_with_ing(verb):
    """
    This function returns True if a given verb string ends with "ing" (case-sensitive),
    and False otherwise.
    """
    # The endswith() method directly returns True or False
    new_verb = verb.endswith('ing')
    return new_verb

# Call the function and print the result
result = ends_with_ing(verb)
print(result)

# Coding Challenge: Conditional Message
# Task: Write a Python function check_keyword(text, keyword) that takes 
# two strings. If text contains keyword (case-insensitive), 
# print "Keyword found!". Otherwise, print "Keyword not found.".
# Example: check_keyword("Learning Python is fun", "python") should 
# print "Keyword found!".
# Get input from the user and convert to lowercase for case-insensitive check
text = input('Enter text: ').lower()
keyword = input('Enter a word to search for: ').lower()

# Define the function
def check_keyword(text, keyword):
    """
    This function checks if a keyword is present in a text (case-insensitive)
    and returns an appropriate message.
    """
    if keyword in text:
        return 'Keyword found!'
    else:
        return 'Keyword not found!'

# Call the function and print the returned message
result = check_keyword(text, keyword)
print(result)

# Coding Challenge: Safe Username Check
# Task: Write a Python function is_safe_username(username) 
# that returns True if the username does NOT contain 
# the characters '%' or '$'. Otherwise, return False.
# Example: is_safe_username("user_name") should return 
# True. is_safe_username("user%name") should return False.
# Get input from the user
user_name = input('Enter username: ')

# Define the function
def is_safe_username(user_name):
    """
    This function returns True if the username does NOT contain
    the characters '%' or '$'. Otherwise, returns False.
    """
    if '%' not in user_name and '$' not in user_name:
        return True # Return actual boolean True
    else:
        return False # Return actual boolean False

# Call the function and print the result
result = is_safe_username(user_name)
print(f"Is '{user_name}' a safe username? {result}")

# Coding Challenge: Standardize Phone Number
# Task: Write a Python function format_phone_number(number_string) 
# that takes a string number_string and removes all hyphens - from it.
# Example: format_phone_number("123-456-7890") should return "1234567890".
# Get input from the user
phone_num = input('Enter a phone #: ')

# Define the function
def format_phone_number(phone_num):
    """
    This function removes hyphens from a phone number string.
    """
    new_phone = phone_num.replace('-','')
    return new_phone

# Call the function and print the result
result = format_phone_number(phone_num)
print(f"Cleaned phone number: {result}")

# Part 3: Slicing & Advanced Substring Extraction
# Coding Challenge: Get First 3 Characters

# Task: Write a Python function get_first_three(s) that takes a string s 
# and returns its first three characters. Assume s will always be 
# at least 3 characters long.
# Example: get_first_three("Elephant") should return "Ele".
# Coding Challenge: Get Last 2 Characters
# Get input from the user
s = input('Enter a string: ')

# Define the function
def get_first_three(s):
    """
    This function takes a string s and returns its first three characters.
    Assumes s will always be at least 3 characters long.
    """
    new_s = s[:3] # Slice from the beginning up to (but not including) index 3
    return new_s

# Call the function and print the result
result = get_first_three(s)
print(f"The first three characters are: {result}")

# Task: Write a Python function get_last_two(s) that takes a 
# string s and returns its last two characters. 
# Assume s will always be at least 2 characters long.
# Example: get_last_two("Coding") should return "ng".
# Get input from the user
s = input('Enter a string: ')

# Define the function
def get_last_two(s):
    """
    This function takes a string s and returns its last two characters.
    Assumes s will always be at least 2 characters long.
    """
    new_s = s[-2:] # Slice from the second-to-last character to the end
    return new_s

# Call the function and print the result
result = get_last_two(s)
print(f"The last two characters are: {result}")

# Coding Challenge: Remove First Character
# Task: Write a Python function remove_first_char(s) that takes a 
# string s and returns a new string with its first character removed. 
# Assume s will always be non-empty.
# Example: remove_first_char("apple") should return "pple".
# Get input from the user
s = input('Enter a string: ')

# Define the function
def remove_first_char(s):
    """
    This function takes a string s and returns a new string
    with its first character removed.
    """
    new_word = s[1:] # Slice from index 1 to the end
    return new_word

# Call the function and print the result
result = remove_first_char(s)
print(f"String with first character removed: {result}")

# Coding Challenge: Reverse a Word
# Task: Write a Python function reverse_word(word) that takes 
# a string word and returns a new string with the 
# characters in reverse order. You must use string slicing.
# Example: reverse_word("hello") should return "olleh".
# Get input from the user
word = input('Enter a word: ')

# Define the function
def reverse_word(word):
    """
    This function takes a string word and returns a new string
    with the characters in reverse order using string slicing.
    """
    rev_word = word[::-1] # Slicing with a step of -1 reverses the string
    return rev_word

# Call the function and print the result
result = reverse_word(word)
print(f"The reversed word is: {result}")

# Coding Challenge: Every Other Letter
# Task: Write a Python function every_other_letter(text) 
# that returns a new string containing every other character 
# from text, starting with the first character.
# Example: every_other_letter("programming") should return "pormn".
# Get input from the user
text = input('Enter a word: ')

# Define the function
def every_other_letter(text):
    """
    This function returns a new string containing every other character
    from text, starting with the first character, using string slicing.
    """
    new_text = text[::2] # Slice with a step of 2
    return new_text

# Call the function and print the result
result = every_other_letter(text)
print(f"Every other letter: {result}")

# Coding Challenge: Extract Domain from Email
# Task: Write a Python function get_domain(email) that takes an 
# email string and extracts the domain name (the part after the '@' symbol). 
# Assume the email will always contain exactly one '@' symbol.
# Example: get_domain("user@example.com") should return "example.com".
# Get input from the user
email = input('Enter an email: ')

# Define the function
def get_domain(email):
    """
    This function takes an email string and extracts the domain name
    (the part after the '@' symbol).
    Assumes the email will always contain exactly one '@' symbol.
    """
    at_index = email.find('@') # Find the index of the '@' symbol
    domain = email[at_index + 1:] # Slice from one character after '@' to the end
    return domain

# Call the function and print the result
result = get_domain(email)
print(f"The domain is: {result}")

# Part 4: Quiz - Predict the Output!
# For each of the following, write down what you think 
# the Python code will print or evaluate to.

# Quiz: Indexing Basics
# Python
# s = "Keyboard"
# print(s[0])
# print(s[5])
# print(s[-1])
# print(s[-len(s)])
# Quiz: Slicing Basics
# KadK

# Python
# text = "Technology"
# print(text[2:6])
# print(text[:4])
# print(text[4:])
# print(text[::3])
# print(text[1:7:2])
# Quiz: Method Return Values
# chnol
# tech
# nology
# thly
# eh0

# Python
# phrase = "Learning Python is fun"
# print(len(phrase))
# print(phrase.find("Python"))
# print(phrase.find("java"))
# print("Python" in phrase)
# print(phrase.upper())
# print("HELLO".lower())
# Quiz: replace() and count()

# Python
# data = "apple,banana,cherry,apple"
# print(data.replace("apple", "orange"))
# print(data.count("apple"))
# print(data.replace("a", "A", 2)) # Replace only first 2 'a's
# Quiz: startswith() / endswith()

# Python
# filename = "document.pdf"
# print(filename.startswith("doc"))
# print(filename.endswith(".txt"))
# print(filename.endswith("pdf"))
# Quiz: Chained Methods

# Python
# message = "  Hello World!  "
# print(message.strip().lower().replace("world", "everyone"))
# Quiz: Tricky Slicing

# Python
# data = "abcdefg"
# print(data[::-1]) # Reversing
# print(data[2:-2]) # Inner part
# print(data[len(data)//2:]) # From middle to end
```
