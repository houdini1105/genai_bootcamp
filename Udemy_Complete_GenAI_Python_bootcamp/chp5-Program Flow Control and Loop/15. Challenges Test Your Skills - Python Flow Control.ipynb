```python
# Challenge #1

# Challenge #1

# Write a Python script that asks the user for a number and prints a
# list of all its divisors for each number less than the given number.

user_num = int(input('Enter a #: '))

# Loop through each number from 1 up to (but not including) user_num
for num in range(1, user_num):
    # Initialize an empty list to store divisors for the current 'num'
    divisors = []

    # Inner loop to check for potential divisors from 1 up to 'num'
    for i in range(1, num + 1):
        # Check if 'i' divides 'num' evenly
        if num % i == 0:
            # If it does, add 'i' to the divisors list
            divisors.append(i)

    # After checking all potential divisors for 'num', print the result
    print(f"Divisors of {num}: {divisors}")

# Challenge #2
# Write a Python program to check if an integer x is a power of
# another integer y. Prompt the user to input both numbers.
# Input: 16, 2
# Output: 2 ** 4 = 16

# Solution from course
x = int(input("Enter a number x: "))
y = int(input(f"Enter a number y to test if x which is {x} is a power of y: "))

found = False # Flag to indicate if a power was found

# Loop through possible exponents (k) starting from 2 up to x // 2
for k in range(2, x//2): 
    if y ** k == x: # Directly checks if y to the power of k equals x
        print(f"{y} ** {k} = {x}")
        found = True # Set flag to True
        break        # Exit the loop early because we found it
# This 'else' block executes ONLY if the 'for' loop finishes naturally (i.e., no 'break' was encountered)
else: 
    print(f'{x} is not the power of {y}')
# Solution using Gemini
x = int(input('Enter a 1st #: '))
y = int(input('Enter 2nd #: '))

counter = 0  # To count the exponent
original_x = x # Store the original value of x before it gets modified

# Handle edge cases first
if x == 1:
    # 1 is y to the power of 0, for any non-zero y
    print(f'Output: {y} ** 0 = 1')
elif y == 0:
    # Cannot divide by zero
    print('Cannot divide by 0')
elif y == 1:
    # If y is 1, x can only be a power of y if x is also 1 (handled above)
    print(f'{original_x} is not a power of {y}')
else:
    # Main logic: Repeatedly divide x by y as long as it's perfectly divisible
    while x % y == 0:
        x //= y  # Integer division to reduce x
        counter += 1 # Increment the exponent counter

    # After the loop, check if x has been reduced to 1
    if x == 1:
        # If x is 1, it means original_x was a perfect power of y
        print(f'Output: {y} ** {counter} = {original_x}')
    else:
        # If x is not 1, original_x was not a perfect power of y
        print(f'{original_x} is not a power of {y}')

# Challenge #3
# Write a Python program that counts and displays the vowels of a given string, ignoring the letter case.
# Input str: Hello Everybody!
# Output: 5
sent = input('Enter a sentence:').lower()
vowels = ('a', 'e', 'i', 'o', 'u')
count = 0
for char in sent:
    if char in vowels:
        count += 1
print(F'String # of vowels {count}')

# Challenge #4
# Write a Python script that checks whether a triangle is equilateral, 
# isosceles, or scalene.
# Prompt the user to enter the lengths of the three sides.
# Triangle Types:
# Equilateral: All three sides are equal.
# Isosceles: Two sides are equal.
# Scalene: All sides are different.
# Input: Enter the lengths of the triangle sides:

# x: 6

# y: 8

# z: 12

# Output: Scalene triangle.
x = int(input('Enter a num for x:'))
y = int(input('Enter a num for y:'))
z = int(input('Enter a num for z:'))
if x + y > z and x + z > y and y + z > x:
    if x == y and y==z and z==x:
        print('Equilateral Triangle')
    elif x == y or y == z or z == x:
        print('Isosceles Triangle')
    else:
        print('Scalene Triangle')


# Challenge #5
# Write a Python program that prompts the user for multiple float numbers and 
# calculates:

# The sum
# The product
# The average
# Enter 0 to finish.


# Challenge #6
# Given a string, write a program that calculates the sum and average 
# of all digits in the string, ignoring other characters.
# Example:
# Input: "Python31py50"
# Output: Sum: 9, Average: 2.25


# Challenge #7
# Write a Python program that displays the multiplication table 
# (from 1 to 10) for a number entered by the user.
# Input: User enters 8
# Output:

# 8 x 1 = 8

# 8 x 2 = 16

# 8 x 3 = 24

# 8 x 4 = 32

# 8 x 5 = 40

# 8 x 6 = 48

# 8 x 7 = 56

# 8 x 8 = 64

# 8 x 9 = 72

# 8 x 10 = 80


# Challenge #8
# Write a Python script that displays the following pattern from 1  to n 
# where n is entered by the user.
# If the user enters 6 it will display:

# 1

# 22

# 333

# 4444

# 55555

# 666666


# Challenge #9
# Write a Python program that finds the common characters that appear 
# in two given strings.


# Challenge #10

# Write a Python program that iterates through numbers from 1 to 50 and prints:
# "Foo" for multiples of 3
# "Bar" for multiples of 5
# "FooBar" for multiples of both 3 and 5


# Challenge #11
# Write a Python script that prints out the Fibonacci series up to a 
# given number n.
# Fibonacci Series: https://www.mathsisfun.com/numbers/fibonacci-sequence.html
# Example: if n is 23 it will print out 0, 1, 1, 2, 3, 5, 8, 13, 21

# Challenge #12
# Write a Python script that draws the following pattern using for loops.

# *

# * *

# * * *

# * * * *

# * * * * *

# * * * *

# * * *

# * *
```
