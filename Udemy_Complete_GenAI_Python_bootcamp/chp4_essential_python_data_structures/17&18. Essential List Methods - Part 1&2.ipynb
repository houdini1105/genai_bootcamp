```python
# Chp4 Ep17&18. Essential List Methods - Part 1 & 2
l1 = list() # using list constructor
#print(dir(l1)) # This will show all list methods
# Methods with _ _ are special methods 

# Adding elements to a list: 
# append() adds a single element to the end of the list
# Cannot and multiple items to a list using append, it will return a error,
# however you can add a list as a single item
l1 = [1, 2.2, 'abc']
l1.append(5) # single element added to the list
l1.append(['x','y']) # list added as a single element
print(l1)

# extend() extends the list by adding each element from an iterable
l1.extend(['s', 'u'])
print(l1)

# insert lets you add a item at a specific poistion via the index
years = [2021, 2022, 2024]
years.insert(2, 2023) # insert needs the index # with the element inserting
print(years)

# removing elements from a list:
# clear() empties the list entire
l1 = ['q', 'b', 'z']
l1.clear() # use clear() to empty the entire list
print(l1)
# pop() removes and returns a eleement at a specific index
# BY default it removes the last index by default if none are provided
l2 = [1, 2, 3, 4]
x = l2.pop() # removes the last index by default
print(x, l2)
# pop() using an index
y = l2.pop(1)
print(y, l2)
# remove() deletes the first occurence of a specific element
# remove() will remove by value instead of by index
l3 = [10, 20, 10, 40]
l3.remove(10)
print(l3)
# To remove all occurneces of an element use a while loop
l4 = [10, 20, 10, 40]
while 10 in l4:
    l4.remove(10)
print(l4)

# Chp4 Ep18. Essential List Methods - Part 2
# list.count()
letters = list('abaca') # using list constructor
print(letters)
print(letters.count('a'))
# list.reverse()
seq = [1, 2, 3, 4]
seq.reverse()
print(seq)
# list.sort(), sorted(list)
# list.sort() is a in place sorting method that orders the elements the order of the list permently
# it modifies the original list
nums = [3, 1, 4, 1, 5]
nums.sort()
print(nums)
# you can also sort in decsending order by placing reverse = True
nums.sort(reverse=True)
print(nums)
# sorted(list) a built in function that returns a new list sorted and leaving the org unchanged
nums = [3, 1, 4, 1, 5]
sorted_nums = sorted(nums) # this is not a method but a built in function
sorted_nums = sorted(nums, reverse=True)
# it does not change the org list but returns a new list
print(nums, sorted_nums)
```
