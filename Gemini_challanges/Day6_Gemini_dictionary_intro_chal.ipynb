```python
# Python Dictionary Operations & Methods Challenges (Beginner Level)

# Challenge 1: Understanding Assignment vs. Copying (Part 1)
# Create a dictionary called original_settings with at least two 
# key-value pairs (e.g., 'theme': 'dark', 'font_size': 14). 
# Then, create a new variable assigned_settings and assign 
# original_settings to it. Change the value of one key in original_settings. 
# Print both original_settings and assigned_settings to observe the effect.
original_settings = {'theme': 'dark', 'font_size': 14}
assigned_settings = original_settings
original_settings['theme'] = 'light'
print(original_settings, assigned_settings)
# Excellent! You've correctly demonstrated that when you assign one 
# dictionary to another without using .copy(), both variables refer to the 
# same dictionary in memory. Changing one also changes the other.

# Challenge 2: Understanding Assignment vs. Copying (Part 2)
# Using the original_settings dictionary from Challenge 1, 
# create a new variable copied_settings using the .copy() method. Now, 
# change the value of a different key in original_settings. 
# Print both original_settings and copied_settings to observe the 
# difference compared to Challenge 1.
original_settings = {'theme': 'dark', 'font_size': 14}
copied_settings = original_settings.copy()
original_settings['theme'] = 'batman'
print(original_settings, copied_settings)
# Excellent! You've perfectly demonstrated the use of the .copy() method. 
# As you can see, when original_settings['theme'] was changed to 'batman', 
# copied_settings remained unaffected, proving that copy() 
# creates an independent duplicate of the dictionary.

# Challenge 3: Merging Dictionaries with .update()
# Create two dictionaries: user_info = {'name': 'Alice', 'email': 'alice@example.com'} and 
# preferences = {'notifications': True, 'theme': 'light'}. 
# Use the .update() method to merge preferences into user_info. 
# Print the user_info dictionary after the merge.
user_info = {'name': 'Alice', 'email': 'alice@example.com'}
preferences = {'notifications': True, 'theme': 'light'}
user_info.update(preferences)
print(user_info)

# Challenge 4: Overwriting with .update()
# Create a dictionary default_config = {'timeout': 30, 'retries': 3}. 
# Create another dictionary user_override = {'retries': 5, 'log_level': 'DEBUG'}. 
# Use .update() to apply user_override to default_config. 
# Print default_config to see how retries was updated and log_level was added.
default_config = {'timeout': 30, 'retries': 3}
user_override = {'retries': 5, 'log_level': 'DEBUG'}
default_config.update(user_override)
print(default_config)

# Challenge 5: Clearing a Dictionary
# Create a dictionary temp_data = {'item1': 100, 'item2': 200}. 
# Use the .clear() method to empty the dictionary. 
# Print temp_data to confirm it's empty.
temp_data = {'item1': 100, 'item2': 200}
print(temp_data)
temp_data.clear()
print(temp_data)

# Challenge 6: Accessing All Keys
# Using the base_config dictionary from the reference material 
# (after it's been updated by version_config):

# Python

base_config = {'batch_size' : 32, 'epochs' : 10}
version_config = {'learning_rage' : 0.001, 'units': 128}
base_config.update(version_config)
# # base_config should now be: {'batch_size': 32, 'epochs': 10, 'learning_rage': 0.001, 'units': 128}
# Print all the keys of base_config using the .keys() method.
print(base_config.keys())

# Challenge 7: Accessing All Values
# Using the same base_config dictionary from Challenge 6, 
# print all the values using the .values() method.
base_config = {'batch_size' : 32, 'epochs' : 10}
version_config = {'learning_rage' : 0.001, 'units': 128}
base_config.update(version_config)
# print all the values using the .values() method.
print(base_config.values())

# Challenge 8: Iterating Through Key-Value Pairs
# Using the base_config dictionary from Challenge 6, 
# write a for loop that iterates through its key-value
#  pairs using .items(). Inside the loop, print each key and
#  its corresponding value in the format: Key: [key], Value: [value].
base_config = {'batch_size' : 32, 'epochs' : 10}
for key, value in base_config.items():
    print(f'Key: {key}, Value: {value}')

# Challenge 9: Checking Key Membership
# Create a dictionary inventory = {'apple': 50, 'banana': 30, 'orange': 20}. 
# Check if the key 'banana' exists in inventory and 
# print True or False. Then, check if 'grape' exists and print the result.
inventory = {'apple': 50, 'banana': 30, 'orange': 20}
print('banana' in inventory)
print('grape' in inventory)

# Challenge 10: Checking Value Membership
# Using the inventory dictionary from Challenge 9, 
# check if the value 30 exists among its values and print True or False. 
# Then, check if 100 exists and print the result.
inventory = {'apple': 50, 'banana': 30, 'orange': 20}
print(30 in inventory.values())
print(100 in inventory.values())

# Challenge 11: Checking Item Membership
# Using the inventory dictionary from Challenge 9, 
# check if the key-value pair ('apple', 50) exists in inventory and 
# print True or False. Then, check for ('banana', 40) and print the result.
inventory = {'apple': 50, 'banana': 30, 'orange': 20}
print(('apple', 50) in inventory.items())
print(('banana', 40) in inventory.items())

# Challenge 12: Combining .get() and .update()
# Create an empty dictionary user_settings = {}. Prompt the user to 
# enter their username and email. Use .update() to add these to user_settings. 
# Then, try to get a theme setting using .get(), providing a 
# default of 'light' if it's not present. Print user_settings and the 
# retrieved theme.
userid = input('Enter your userid:')
email = input('Enter your email:')
user_settings = {}
user_settings.update({'username' : userid, 'email' : email})
theme = user_settings.get('theme', 'light')
print(user_settings, theme)

# Challenge 12b: Product Price List

# Create an empty dictionary called product_prices = {}.
# Ask the user to enter the names and prices for 4 different products.
# Add these product names (as keys) and their 
# prices (as values) to the product_prices dictionary. Make sure to store 
# the prices as floating-point numbers (e.g., 12.99, 5.00).

# After collecting all 4 entries, iterate through the product_prices dictionary 
# using .items() and print each product's name and price in the format:
# Product: [product_name] - Price: $[product_price]
product_prices = {}
for x in range(4):
    print(f"\nEntering details for product {x + 1}:") # Good practice to show progress
    product = input('Enter product:')
    price = float(input('Enter Amount:'))
    product_prices[product] = price
for k, v in product_prices.items():
    print(f'Product: {k} - Price: ${v}')

# Challenge 13: Dynamic Dictionary Creation and Iteration
# Create an empty dictionary student_grades = {}. Ask the user 
# to enter names and grades for 3 students. Add these as key-value pairs to 
# student_grades. After collecting all entries, iterate through student_grades 
# using .items() and print each student's name and grade.
#student_name = input("Enter student's name: ")
# You might want to convert the grade to an integer or float if you plan to do calculations later
#student_grade = int(input(f"Enter {student_name}'s grade: "))
student_grades = {} # Start with your empty dictionary

# This loop will run 3 times (once for each student)
for i in range(3):
    print(f"\nEntering data for student {i + 1}:") # Just to show which student we're on

    # Step 1: Get the student's name
    name = input("Enter student's name: ")

    # Step 2: Get the student's grade
    # We use int() here because grades are usually numbers, so we convert the input string to an integer
    grade = int(input(f"Enter {name}'s grade: "))

    # Step 3: !!! YOU COMPLETE THIS PART !!!
    # Add the 'name' as a key and 'grade' as its value to the 'student_grades' dictionary.
    # Remember how you added to a dictionary in previous challenges using [] = value?
    student_grades[name] = grade
for k, v in student_grades.items():
    print(f'{k} -> {v}')


# Challenge 14: Counting Occurrences (Using Keys/Values)
# Given a list of fruits: fruits = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']. 
# Create an empty dictionary fruit_counts = {}. Iterate through the fruits list. 
# For each fruit, if it's already a key in fruit_counts, increment its value. Otherwise, 
# add it as a new key with a value of 1. Print fruit_counts at the end.
fruits = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
fruit_counts = {}
for x in fruits:
    if x in fruit_counts:
        fruit_counts[x] += 1
    else:
        fruit_counts[x] = 1
for k, v in fruit_counts.items():
    print(fruit_counts)

# Challenge 15: Deleting a Key (Bonus)
# (This operation wasn't explicitly covered but is a 
# common dictionary operation.) Using the inventory dictionary 
# from Challenge 9, remove the 'orange' entry. Print the 
# inventory dictionary after removal. (Hint: Look up how to del a key from 
# a dictionary).
inventory = {'apple': 50, 'banana': 30, 'orange': 20}
removed_value = inventory.pop('orange')
print(inventory, removed_value)
```
