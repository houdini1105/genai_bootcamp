```python
# Challenge 1: Basic Function Creation
# Create a Python function called greet_user that takes one 
# argument, name, and prints a greeting message like "Hello, [name]!".
def greet_user(name):
    print(f'Hello, {name}!')

print(greet_user('Ted'))

# Challenge 2: Function with Multiple Positional Arguments
# Write a function called calculate_area that takes two positional
# arguments, length and width, and returns the area of a rectangle. Then,
# call the function with example values and print the result.
def calculate_area(length, width):
    area = length * width
    return area

# Call the function with example values and print the result
print(calculate_area(2, 1))
print(calculate_area(5, 10))

# Challenge 3: Function with Default Arguments
# Design a function named set_volume that takes two arguments: device and level. 
# The level argument should have a default value of 50. The function 
# should print a message like "Setting volume on [device] to [level]%". Test 
# it by calling the function with and without providing a level.
def set_volume(device, level=50):
    print(f'Setting volume on {device} to {level}%')

set_volume('TV')
set_volume('TV', 75)

# Challenge 4: Overriding Default Arguments
# Modify the set_volume function from Challenge 3. Call it again, 
# but this time, explicitly set the level to 80 for a specific device.
def set_volume(device, level=50):
    print(f'Setting volume on {device} to {level}%')

set_volume('TV', 80)

# Challenge 5: Combining Positional and Keyword Arguments
# Create a function display_product_info that takes product_name (positional) 
# and price (keyword argument with a default of 10.0). The function should 
# print a formatted string showing the product name and its price. 
# Demonstrate calling it with both positional and keyword arguments.
def display_product_info(product_name, price=10.0):
    print(f'The {product_name} price is ${price}.')

display_product_info(product_name='JBL', price=10.99)
display_product_info('sony', 8.99)

# Challenge 6: Understanding Positional-Only Arguments
# Write a function multiply_numbers that takes two positional-only 
# arguments, x and y, and returns their product. Use the / syntax to 
# enforce positional-only arguments. Try calling it with keyword arguments 
# and observe the error.
def multiply_numbers(num1, num2, /):
    sum = num1 * num2
    return sum

multiply_numbers(2, 3)
multiply_numbers(num1=2, num2 = 3)

# Challenge 7: Function with a Mix of Argument Types
# Define a function configure_settings that takes:
# setting_name (positional-only)
# value (can be positional or keyword)
# debug_mode (keyword-only with a default of False)
# The function should print a message indicating the configuration.
def configure_settings(setting_name, /, value, *, debug_mode=False):
    print(f'{setting_name}, {value}, {debug_mode}')

configure_settings('Network Timeout', 300)

# Challenge 8: Using *args for Variable Positional Arguments
# Create a function sum_all that accepts any number of positional arguments 
# using *args. The function should return the sum of all the numbers 
# passed to it. Test it with various numbers of arguments.
def sum_all(*args):
    print(args)
    return sum(args) # sum is a built-in Python functionused to add up all 
#the numbers in an iterable (like a list, tuple, or set) and return their total.

sum_all(5, 3, 4)

# Challenge 9: Finding the Maximum with *args
# Write a function find_max that takes a variable number of positional 
# arguments using *args. The function should return the largest number 
# among the arguments.
def find_max(*args):
    print(args)
    return max(args)

find_max(2, 5, 9, 4, 1, 10)

# Challenge 10: Using **kwargs for Variable Keyword Arguments
# Develop a function print_user_profile that accepts any number of keyword 
# arguments using **kwargs. The function should iterate through the kwargs 
# and print each key-value pair, representing user profile information.
def print_user_profile(**kwargs):
    for key, value in kwargs.items():
        print(f'{key} --> {value}')

print_user_profile(printer='Hp4200', duplex = True)

# Challenge 11: Combining *args and **kwargs
# Create a function process_data that takes some initial positional 
# arguments, then *args, and finally **kwargs. Inside the function, print the 
# initial arguments, the args tuple, and the kwargs dictionary.
def process_data(num, *args, **kwargs):
    print(f'{num}, {args}, {kwargs}')

process_data(7, 1, 2, 3, 4, item = '3d Printer')

# Challenge 12: Basic Function Return Value
# Write a function is_even that takes an integer as input and 
# returns True if the number is even, and False otherwise.

def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False

is_even(8)

# Challenge 13: Multiple Return Values
# Create a function get_stats that takes a list of numbers as input. 
# It should return two values: the sum of the numbers and their average.
def get_stats(nums):
    total = sum(nums)
    length = len(nums)
    if length == 0:
        return 0, None
    else:
        avg = total / length
        return total, avg
    
get_stats([5, 10, 25])

# Challenge 14: Handling Division by Zero (Error Prevention)
# Modify the divide function from the reference material. Instead of raising 
# a ValueError, make it return None if the denominator is 0, and the 
# result of the division otherwise.
def divide(numerator, denominator, /): # / both parms to be postional=only
    """Devides two numbers"""
    if denominator == 0:
        return None
    else:
        return numerator / denominator

# correct usage
result = divide(10, 2)
print(result)

# Challenge 15: Applying Positional-Only and Default Arguments
# Design a function create_log_entry that takes:
# message (positional-only)
# level (with a default value of 'INFO')
# The function should print a formatted log entry including the message 
# and its level. Demonstrate calling it by only providing the message, and 
# then by providing both the message and a different level.
def create_log_entry(message, /, level='Info'):
    print(f'{message}, {level}')

create_log_entry('Test')
create_log_entry('Test', level="High")
```
