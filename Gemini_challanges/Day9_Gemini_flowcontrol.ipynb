```python
# Challenge #13
# Write a Python script that takes a string as input from the user
# and counts the number of uppercase letters, lowercase letters, and
# digits in the string.
# Example:
# Input: "Python3.9"
# Output:
# Uppercase: 1
# Lowercase: 6
# Digits: 2

sent = input('Enter a sent:') # Get the string input
upper = 0                  # Initialize counter for uppercase letters
lower = 0                  # Initialize counter for lowercase letters
digits = 0                 # Initialize counter for digits

# Iterate through each character in the input string
for char in sent:
    if char.isupper():   # Check if the character is an uppercase letter
        upper += 1
    elif char.islower(): # Check if the character is a lowercase letter
        lower += 1
    elif char.isdigit(): # Check if the character is a digit
        digits += 1

# Print the final counts
print(f'Uppercase: {upper}')
print(f'Lowercase: {lower}')
print(f'Digits: {digits}')


# Challenge #14
# Write a Python program that iterates through numbers from 1 to 20 and prints:
# "Buzz" for multiples of 3
# "Fizz" for multiples of 5
# "FizzBuzz" for multiples of both 3 and 5
# The number itself for all other cases.
for x in range(1, 21):
    if x % 3 == 0 and x % 5 == 0:
        print('FizzBuzz')
    elif x % 3 == 0:
        print('Buzz')
    elif x % 5 == 0:
        print('Fizz')
    else:
        print(f'{x}')

# Challenge #15
# Write a Python script that prints the first n numbers of the Fibonacci series,
# where n is a number entered by the user.
# This time, you will stop after printing n numbers, not when a value limit is reached.
# Example: if n is 8, it will print the first 8 numbers:
# 0, 1, 1, 2, 3, 5, 8, 13

n = int(input('Enter a num:'))
first_num = 0
second_num = 1

if n == 0:
    print('') # Prints an empty line if 0 terms are requested
elif n == 1:
    print('0') # Prints only the first term (0) if 1 term is requested
else: # Handles n >= 2
    # Print the first two terms with appropriate end=''
    print(first_num, end=', ')
    print(second_num, end='')

    # Loop n-2 times to get the remaining terms
    for x in range(n - 2):
        # Calculate the next Fibonacci number
        next_fib = first_num + second_num
        # Print it with a leading comma and space
        print(', ' + str(next_fib), end='')
        # Update first_num and second_num for the next iteration
        first_num = second_num
        second_num = next_fib

# Add a final newline character for clean output, especially if end='' was used
print()

# Challenge #16
# Write a Python program that generates a geometric sequence up to a given
# number n, entered by the user. A geometric sequence starts at 1, and
# each subsequent number is double the previous one.
# Example: If n is 40, it will print:
# 1, 2, 4, 8, 16, 32

n = int(input('Enter a number (n): ')) # Get the limit number from the user
seq_num = 1 # The sequence starts at 1

# Handle edge case where n is less than 1 (no terms to print)
if n < 1:
    print('') # Prints an empty line
else:
    # Always print the first term (1)
    # Use end='' to keep the cursor on the same line
    print(seq_num, end='')

    # Calculate the first potential *next* number
    next_possible_num = seq_num * 2

    # Loop as long as the *next* calculated number is less than or equal to n
    while next_possible_num <= n:
        # If the next_possible_num is valid, update seq_num to this value
        seq_num = next_possible_num
        # Print the current valid seq_num, preceded by a comma and space
        # Use end='' to stay on the same line
        print(f', {seq_num}', end='')

        # Calculate the *next* potential number for the next loop iteration's check
        next_possible_num = seq_num * 2

# After the loop finishes, print a final newline character for clean output
print()

# Challenge #17
# Write a Python script that takes a string of numbers separated by
# commas (e.g., "1,2,3,4,5") as input. The script should then calculate
# and print the sum and average of the numbers, handling any non-digit
# characters gracefully.
# Example:
# Input: "1,2,three,4,5"
# Output:
# Sum: 12
# Average: 3.0
# (Note: The script should ignore the word "three")

# You can uncomment this line to get user input instead of using the hardcoded string
# user_string = input('Enter numbers separated by commas: ')
user_string = "1,2,three,4,5" # Hardcoded string for testing as per the example

split_string = user_string.split(',') # Split the string by commas

total_sum = 0 # Initialize sum
count = 0     # Initialize count of valid numbers

# Iterate through each item (which is a string) in the split list
for item in split_string:
    # Check if the item consists only of digits
    if item.isdigit():
        # If it's a digit, convert it to an integer and add to total_sum
        total_sum += int(item)
        # Increment the count of valid numbers
        count += 1

# After the loop finishes, check if any valid numbers were found
if count > 0:
    # Calculate the average
    average = total_sum / count
    # Print the sum and average in the specified format
    print(f'Sum: {total_sum}, Average: {average}')
else:
    # If no valid numbers were found, print a message
    print("No valid numbers found in the input string.")
```
