```python
# Challenge 1: Basic Function Creation
# Objective: Create a simple function that prints a greeting.
# Challenge: Write a Python function called greet_user that takes no 
# arguments and prints the message "Hello, Python coder!".
def greet_user():
    """This will greet the user"""
    print('Hello, Python coder!')
    return # return does not need to be included but good practice
greet_user()

# Challenge 2: Function with a Parameter
# Objective: Understand how to pass arguments to a function.
# Challenge: Create a function named display_name that accepts one argument, 
# name, and prints "My name is [name]." (replace [name] with the value 
# passed to the function).
def display_name(name):
    """ This will print the name as an argument """
    print(f'My name is {name}')
    return # return does not need to be included but good practice
display_name('Dennis')

# Challenge 3: Using Docstrings
# Objective: Learn to document your functions using docstrings.
# Challenge: Modify the display_name function from Challenge 2 to include a 
# docstring that explains its purpose (e.g., "This function displays a 
# given name."). After defining the function, demonstrate how to access 
# its docstring using both help() and the __doc__ attribute.
def display_name(name):
    """ This will print the name as an argument """
    print(f'My name is {name}')
    return # return does not need to be included but good practice
display_name('Dennis')

help(display_name)
print(display_name.__doc__)

# Challenge 4: Simple Return Statement
# Objective: Understand the basic usage of the return statement.
# Challenge: Write a function called add_two_numbers that takes 
# two parameters, num1 and num2, and returns their sum. Then, call this 
# function with two numbers and print the returned result.
def add_two_numbers(num1, num2):
    sum = num1 + num2
    return sum

print(add_two_numbers(3, 5))

# Challenge 5: return vs. print
# Challenge 5: return vs. print
# Objective: Differentiate between printing within a function and 
# returning a value.
# Challenge: Create two functions:
# print_product(a, b): This function should calculate the product of a and b 
# nd print the result directly inside the function.
# get_product(a, b): This function should calculate the product of a and b and 
# return the result.
# Then, call both functions with the same arguments and observe the 
# difference in how you interact with their results.
def print_product(a, b):
    c = a * b
    print(c)

def get_products(a, b):
    c = a * b
    return c

print_product(2, 3)

print(get_products(2, 3))

# Challenge 6: Functions Returning None
# Objective: Understand when a function implicitly returns None.
# Challenge: Define a function called do_nothing that takes no 
# arguments and has no return statement (you can use pass inside it). Call 
# this function and print the value it returns. Explain why it 
# returns what it does.
def do_nothing():
    pass

do_nothing()
print(do_nothing())
# "A function that does not have an explicit return statement, 
# or has a return statement without a value (like just return), 
# automatically returns None."

# Challenge 7: Multiple Return Values
# Objective: Learn to return multiple values from a function as a tuple.
# Challenge: Create a function calculate_stats that takes a single number n 
# as input. This function should return three 
# values: n itself, n squared, and n cubed. Call the function with a 
# number and print the returned tuple.
def multiple_values(n):
    n = n, n ** 2, n ** 3
    return n

print(multiple_values(5))

# Challenge 8: Unpacking Multiple Return Values
# Objective: Practice unpacking multiple returned values into separate variables.
# Challenge: Using the calculate_stats function from Challenge 7, call the 
# function with a number and then unpack the returned values into three 
# separate variables (e.g., original, squared, cubed). Print each of these 
# variables individually.
def multiple_values(n):
    n = n, n ** 2, n ** 3
    return n
x, y, z = multiple_values(5)
print(x, y, z)

# Challenge 9: Focus and Modularity
# Objective: Emphasize the concept of keeping functions focused on one task.
# Challenge: You need to write code that takes a user's name, converts it to 
# uppercase, and then prints a personalized greeting.
# Instead of doing all of this in one function, create two separate functions:
# to_uppercase(text): This function takes a string and returns its uppercase version.
# personal_greeting(name): This function takes a name (which is expected 
# to be in uppercase) and prints "WELCOME, [UPPERCASE_NAME]!".
# Then, demonstrate how to use these two functions together to achieve the 
# desired outcome.

def to_uppercase(text):
    upper_text = text.upper()
    return upper_text

def personal_greeting(name):
    print(f'Welcome, {name}')

print(personal_greeting(to_uppercase('dennis')))

# Challenge 10: Combining Concepts
# Objective: Apply multiple learned 
# concepts (parameters, return, docstrings, help()) in a single challenge.
# Challenge: Design a function named calculate_area_rectangle that takes 
# length and width as parameters.

# It should have a docstring explaining its purpose 
# (e.g., "Calculates the area of a rectangle.").
# It should calculate the area and return the result.
# If either length or width is negative, the function should return the 
# string "Error: Dimensions cannot be negative."
# After defining the function, do the following:
# Call it with valid dimensions and print the area.
# Call it with invalid dimensions (e.g., a negative value) and print the 
# error message.
# Use help() to display the function's docstring.
def calculate_area_rectangle(l, w):
    """ Calculates the area of a rectangle."""
    if l <= 0 or w <= 0:
        return ('Error. Dimensions cannot be negative.')
    else:
        return l * w

print(calculate_area_rectangle(4, 21))

print(calculate_area_rectangle(-4, 0))

help(calculate_area_rectangle)
```
