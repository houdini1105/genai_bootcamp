Given letters = ['p', 'y', 't', 'h', 'o', 'n', 'p'], use the count() method to find out how many times the letter 'p' appears in the list. Print the count.

Given sequence = [1, 5, 3, 8, 2], use the reverse() method to reverse the order of the elements in sequence in-place. Print sequence.

Challenge 12: Sorting a List In-Place (Ascending)
Given unsorted_nums = [9, 2, 7, 1, 5], use the sort() method to sort unsorted_nums in ascending order. Print unsorted_nums.

Given data_points = [100, 20, 80, 40, 60], use the built-in sorted() function with the appropriate argument to create a new list called desc_sorted_points that contains the elements of data_points in descending order. Print both data_points and desc_sorted_points.

Challenge 16: Combining append() and pop()
Start with items_stack = [10, 20, 30]. Simulate a stack operation: pop() an element, then append() a new element 40. Print items_stack.

Given mixed_list = [1, 'a', 2, 'b', 1, 'c', 1]. Remove the first occurrence of 1. Then, count how many times 1 appears in the modified list. Print the count.

Challenge 18: Inserting and then Reversing
Start with my_items = ['first', 'third']. Insert 'second' at the correct position to make the list ordered alphabetically. Then, reverse the entire list. Print the final my_items.

Challenge 19: extend() with Mixed Types
Given initial_data = [True, False]. Extend this list with the elements from [10, 'hello', 3.14]. Print the initial_data.

Challenge 20: Chaining Operations
Given test_list = [5, 2, 8, 1, 5, 9].

Remove the first occurrence of 5.

Sort the remaining elements in ascending order.

Append the count of 5s (from the original list before any modification) to the end of the list.
Print the final test_list.



```python
# Python List Method Challenges

# Challenge 1: Appending a Single Element
# Create an empty list called my_list. Use the append() method to 
# add the integer 10 to my_list. Print the list.
my_list = []
my_list.append(10) # append add a single element
print(my_list)

# Challenge 2: Appending a List as a Single Element
# Start with my_list = [1, 2, 3]. Use append() to add the list ['a', 'b'] as a single element to my_list. Print my_list.
my_list = [1, 2, 3]
my_list.append(['a', 'b']) # append add a single element which is a list
print(my_list)
# Challenge 3: Extending a List
# Start with fruits = ['apple', 'banana']. Use the extend() method to add the elements 'orange' and 'grape' from a list to fruits. Print fruits.
fruits = ['apple', 'banana']
fruits.extend(['orange', 'grape']) # extend the list adding each element from an inteble
print(fruits)
# Challenge 4: Inserting an Element
# Given the list colors = ['red', 'blue', 'green'], use the insert() method to add 'yellow' at index 1. Print colors.
list_colors = ['red', 'blue', 'green']
list_colors.insert( 1,'yellow')
print(list_colors)
# Challenge 5: Clearing a List
# Create a list numbers = [1, 2, 3, 4, 5]. Use the clear() method to empty the list. Print numbers.
numbers = [1, 2, 3, 4, 5]
numbers.clear()
print(numbers)
# Challenge 6: Popping the Last Element
# Given data = ['A', 'B', 'C', 'D'], use pop() (without an argument) to remove the last element. 
# Store the removed element in a variable removed_item and print both removed_item and data.
data = ['a', 'b', 'c', 'd']
removed_item = data.pop()
print(removed_item, data)
# Challenge 7: Popping an Element by Index
# Given items = ['one', 'two', 'three', 'four'], use pop() to remove the element at index 2. Print the modified items list.
items = ['one', 'two', 'three', 'four']
removed_items = items.pop(2)
print(items, removed_items)
# Challenge 8: Removing the First Occurrence of an Element
# Given my_numbers = [10, 20, 30, 20, 40], use remove() to delete the first occurrence of 20. Print my_numbers.
my_numbers = [10, 20, 30, 20, 40]
my_numbers.remove(20)
print(my_numbers)
# Challenge 9: Removing All Occurrences of an Element
# Given repeated_nums = [5, 10, 5, 15, 5, 20], use a while loop and remove() to delete all occurrences of 5 from the list. Print repeated_nums.
repeated_nums = [5, 10, 5, 15, 5, 20]
while 5 in repeated_nums:
    repeated_nums.remove(5)
print(repeated_nums)

# Challenge 10: Counting Elements
# Given letters = ['p', 'y', 't', 'h', 'o', 'n', 'p'], use the count() method to 
# find out how many times the letter 'p' appears in the list. Print the count.
letters = ['p', 'y', 't', 'h', 'o', 'n', 'p']
print(letters.count('p'))

# Challenge 11: Reversing a List In-Place
# Given sequence = [1, 5, 3, 8, 2], use the reverse() method to reverse the order of the elements in sequence in-place. Print sequence.
sequence = [1, 5, 3, 8, 2]
sequence.reverse()
print(sequence)

# Challenge 12: Sorting a List In-Place (Ascending)
# Given unsorted_nums = [9, 2, 7, 1, 5], use the sort() method to sort unsorted_nums in ascending order. Print unsorted_nums.
unsorted_nums = [9, 2, 7, 1, 5]
unsorted_nums.sort()
print(unsorted_nums)
# Challenge 13: Sorting a List In-Place (Descending)
# Given more_nums = [4, 8, 1, 6, 2], use the sort() method with the appropriate argument to sort more_nums in descending order. 
# Print more_nums.
more_nums = [4, 8, 1, 6, 2]
more_nums.sort(reverse=True)
print(more_nums)
# Challenge 14: Using sorted() for a New Sorted List (Ascending)
# Given original_list = [7, 3, 9, 1, 5], use the built-in sorted() 
# function to create a new list called new_sorted_list that contains the elements of original_list in ascending order. 
# Print both original_list and new_sorted_list.
original_list = [7, 3, 9, 1, 5]
new_sorted_list = sorted(original_list)
print(new_sorted_list)
# Challenge 15: Using sorted() for a New Sorted List (Descending)
# Given data_points = [100, 20, 80, 40, 60], use the built-in sorted() 
# function with the appropriate argument to create a new list called desc_sorted_points 
# that contains the elements of data_points in descending order. Print both data_points and desc_sorted_points.
data_points = [100, 20, 80, 40, 60]
desc_sorted_points = sorted(data_points,reverse=True)
print(desc_sorted_points)
# Challenge 16: Combining append() and pop()
# Start with items_stack = [10, 20, 30]. Simulate a stack operation: pop() an element, 
# then append() a new element 40. Print items_stack.
items_stack = [10, 20, 30]
items_stack.pop(0)
items_stack.append(69)
print(items_stack)
# Challenge 17: Counting After Removal
# Given mixed_list = [1, 'a', 2, 'b', 1, 'c', 1]. Remove the first occurrence of 1. 
# Then, count how many times 1 appears in the modified list. Print the count.
mixed_list = [1, 'a', 2, 'b', 1, 'c', 1]
mixed_list.remove(1)
print(mixed_list.count(1))
# Challenge 18: Inserting and then Reversing
# Start with my_items = ['first', 'third']. Insert 'second' at the correct position to make the list ordered alphabetically. 
# Then, reverse the entire list. Print the final my_items.
my_items = ['first', 'third']
my_items.insert( 1, 'second')
print(my_items)
# Challenge 19: extend() with Mixed Types
# Given initial_data = [True, False]. Extend this list with the elements from [10, 'hello', 3.14]. Print the initial_data.
initial_data = [True, False]
initial_data.extend([10, 'hello', 3.14])
print(initial_data)
# Challenge 20: Chaining Operations
# Given test_list = [5, 2, 8, 1, 5, 9].
# Remove the first occurrence of 5.
# Sort the remaining elements in ascending order.
# Append the count of 5s (from the original list before any modification) to the end of the list.
# Print the final test_list.
test_list = [5, 2, 8, 1, 5, 9]

# Store the original count of 5s before any modifications
original_count_of_fives = test_list.count(5)

# Remove the first occurrence of 5.
test_list.remove(5)

# Sort the remaining elements in ascending order.
test_list.sort()

# Append the count of 5s (from the original list before any modification) to the end of the list.
test_list.append(original_count_of_fives)

print(test_list)
```
