```python
# Challenge #1
# Write a Python function that takes a list as an argument and returns a 
# new list containing only the unique elements from the original list, in the 
# same order.
# Sample input: [1, 2, 3, 3, 3, 3, 4, 5, 1, 3, 5, 5, 5]
# Expected output: [1, 2, 3, 4, 5]

def get_unique_elements(input_list):
    new_list = []
    for x in input_list:
        if x not in new_list:
            new_list.append(x)
    return new_list

# Sample usage:
sample_input = [1, 2, 3, 3, 3, 3, 4, 5, 1, 3, 5, 5, 5]
unique_elements = get_unique_elements(sample_input)
print(f"Original list: {sample_input}")
print(f"Unique elements: {unique_elements}")

# However using a set would be a more efficient method
def get_unique_elements_efficiently(input_list):
    unique_list = []
    seen = set()
    for item in input_list:
        if item not in seen:
            seen.add(item)
            unique_list.append(item)
    return unique_list

# Sample usage:
sample_input = [1, 2, 3, 3, 3, 3, 4, 5, 1, 3, 5, 5, 5]
unique_elements = get_unique_elements_efficiently(sample_input)
print(f"Efficient solution output: {unique_elements}")

# Challenge #2
# Write a Python function to check whether a number is perfect. The function 
# should return True if the number is perfect, and False otherwise.
# ðŸ“˜ Learn more about perfect numbers: Perfect Number â€“ Britannica
def perfect_number(num):
    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i
    if num == total:
        return True
    else:
        return False
    # Didn't include if num is 0 logic
    # However a more pythonic way is below
def perfect_number_concise(num):
    # Perfect numbers must be positive integers
    if num <= 0:
        return False

    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i
            
    return num == total

# This is the course solution
def all_divisors(n):
    """
     This function returns all divisors of a number
    """
    divisors = []
    for x in range(1, int(n/2)+1):
        if n % x == 0:
            divisors.append(x)
    return divisors


def perfect_number(n):
    divs = all_divisors(n)
    if sum(divs) == n:
        return True
    else:
        return False


# calling the function
n = 496
if perfect_number(n):
    print(f'{n} is one rare of a kind number, it\'s a perfect number')
else:
    print(f'Nothing special about {n}, it\'s no perfect number')


# Challenge #3
# Write a Python function that takes an integer n as an argument and returns 
# its factorial.
# ðŸ“˜ Learn more about factorials: Factorial â€“ Wikipedia
def factorial(n):
    if n < 0:
        return False  # Or raise an error, since factorials are undefined for negative numbers
    elif n == 0:
        return 1
    
    product = 1
    for i in range(1, n + 1):
        product *= i
        
    return product

# Sample usage:
print(f"The factorial of 5 is: {factorial(5)}")   # Expected output: 120 (5*4*3*2*1)
print(f"The factorial of 0 is: {factorial(0)}")   # Expected output: 1
print(f"The factorial of -3 is: {factorial(-3)}") # Expected output: False

# Gemini solution for recursive
def factorial_recursive(n):
    if n < 0:
        return False
    elif n == 0:
        return 1
    else:
        return n * factorial_recursive(n - 1)

# Sample usage:
print(f"The recursive factorial of 5 is: {factorial_recursive(5)}")

# Course solution
def fact(n):
    if n < 0:
        print('The function\'s argument (n) should be greater than or equal to zero.')
    elif n == 0:
        return 1
    else:
        f = 1
        while n > 0:
            f *= n
            n -= 1
        return f


print(fact(10))


# Challenge #4
# Create a function that takes an integer as an argument and returns True 
# if itâ€™s a prime number, and False otherwise.
def prime_check(n):
    # Prime numbers must be greater than 1
    if n <= 1:
        return False
    
    # Check for divisors from 2 up to n-1
    for i in range(2, n):
        if n % i == 0:
            # If a divisor is found, it's not prime
            return False
    
    # If the loop finishes without finding a divisor, it's prime
    return True

# Sample usage:
print(f"Is 13 a prime number? {prime_check(13)}") # Expected output: True
print(f"Is 15 a prime number? {prime_check(15)}") # Expected output: False (divisible by 3 and 5)
print(f"Is 2 a prime number? {prime_check(2)}")   # Expected output: True
print(f"Is 1 a prime number? {prime_check(1)}")   # Expected output: False

# Gemini optimized solution (need to import math)
import math

def prime_check_optimized(n):
    if n <= 1:
        return False
    
    # Check for divisors up to the square root of n
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    
    return True

# Sample usage:
print(f"Is 97 a prime number? {prime_check_optimized(97)}") # Expected output: True

# Course solution
# Create a function that takes an integer as argument and returns True if itâ€™s a prime number and false otherwise.

def is_prime(n):
    prime = True
    if n == 1: # 1 is not a prime number, the first prime numnber is 2
        return False
    i = 1
    while i < n // 2:
        i = i + 1
        if n % i == 0:
            prime = False
            break
    return prime # returns True or False
        
        


# Challenge #5
# Using the function defined in the previous challenge, find and return five 
# prime numbers greater than 1,000,000.
def prime_check(n):
    # Prime numbers must be greater than 1
    if n <= 1:
        return False
    
    # Check for divisors from 2 up to n-1
    for i in range(2, n):
        if n % i == 0:
            # If a divisor is found, it's not prime
            return False
    
    # If the loop finishes without finding a divisor, it's prime
    return True

def five_prime_nums():
    primes = []
    num_to_check = 1000001
    while len(primes) < 5:
        is_prim = prime_check(num_to_check)
        if is_prim: # This is a boolean check
            primes.append(num_to_check)
        num_to_check += 1
    return primes
        


# Challenge #6
# Write a function called fibo that takes an integer greater than 10 and returns 
# the Fibonacci sequence from 0 up to (but not exceeding) the given number.
# ðŸ“˜ Learn more about the Fibonacci sequence: Fibonacci â€“ Math is Fun
# Example:
# fibo(23) â†’ [0, 1, 1, 2, 3, 5, 8, 13, 21]


# Challenge #7
# Write a function that takes a list of numbers and returns the Equilibrium Index â€” the index where the sum of elements on the left is equal to the sum on the right. Return False if no such index exists.
# ðŸ“˜ Learn more about equilibrium index: Equilibrium Index â€“ GeeksforGeeks


# Challenge #8
# Modify the previous solution so the function accepts a string of numbers separated 
# by commas instead of a list.
# Example:

# nums = '2, 3, 10, 5'
# print(equilibrium_index(nums)) # Output: 2
# nums = '3, 3, 10, 5'
# print(equilibrium_index(nums)) # Output: False
# [Stuck? Click here to see the solution.]



# Challenge #9
# Define a function that draws a Christmas tree using asterisks (*). The function should take a single integer argument representing the tree's height.
# Example:
# draw_tree(5) should output:

#       *

#     ***

#    *****

#   *******

# *********
```
