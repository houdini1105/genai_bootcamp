```python
# # Coding Challenges: For Loops and Ranges
#1. Count Down: Write a program that uses a for loop and the range() 
# function to print a countdown from 5 to 1, and then prints "Go!".
for countdown in range(5, 0, -1):
    print(f'{countdown}')

print('GO!')

#2. Even Numbers: Create a list of numbers from 1 to 20. Then, use a 
# for loop to iterate through the list and print only the even numbers.
numbers = list(range(1, 21))
for even in numbers:
    if even % 2 == 0:
        print(even)

#3. String Reversal: Ask the user to enter a word. Use a for loop to 
# iterate through the word in reverse order and print each character.
word = input('Enter a word:')
for charc in word[::-1]:
    print(charc)

#4. Sum of List Elements: Given a list of numbers, write a for 
# loop to calculate and print the sum of all the numbers in the list.
numbers = list(range(1,5))
total_sum = 0
for num in numbers:
    total_sum += num

print(total_sum)

#4b. Product of List Elements
# Challenge: Given a list of numbers, write a for loop to calculate and 
# print the product (multiplication) of all the numbers in the list.
# For example, if the list is [1, 2, 3, 4], the product would 
# be 1×2×3×4=24.
num_list = [1, 2, 3, 4]
sum_of_nums = 1
for num in num_list:
    sum_of_nums *= num

print(sum_of_nums)

#5. Filtering Words: You have a list of 
# words: ['apple', 'banana', 'grape', 'kiwi', 'orange']. Use a for loop to 
# print only the words that have more than 5 letters.
words = ['apple', 'banana', 'grape', 'kiwi', 'orange']
for word in words:
    if len(word) > 5: 
        print(word)

#6. Multiplication Table: Ask the user to enter a number. Use a for 
# loop and the range() function to print the multiplication table for that 
# number from 1 to 10.
user_number = int(input('Enter a number 1-10:'))
for num in range(1, 11):
    product = user_number * num
    print(f'{user_number} X {num} = {product}')
    # Alternatively, without a 'product' variable:
    # print(f'{user_number} x {num} = {user_number * num}')

#7. Character Counter: Ask the user to enter a sentence. Use a for 
# loop to count how many times the letter 'a' (case-insensitive) appears in 
# the sentence and print the count.
sent = input('Enter a sent:')
counter = 0
for char in sent.lower():
    if char == 'a':
        counter += 1

print(f'{counter}')

#8. List of Squares: Use a for loop and the range() function to create a 
# new list containing the squares of numbers from 1 to 7. Print the new 
# list.
empty_list = []
for num in range(1, 8):
    new_list = num ** 2
    empty_list.append(new_list)
print(empty_list)

#9. Identifying Vowels: Given a word, use a for loop to iterate through 
# its characters. If a character is a vowel (a, e, i, o, u, case-insensitive), 
# print "Vowel found: [character]".
word = input('Enter word;').lower()
vowel = ('a', 'e', 'i', 'o', 'u')
for char in word:
    if char in vowel:
        print(f'Vowel Found {char}')

#10. Repeated Message: Use a for loop with a "throwaway" variable 
# (underscore) to print the message "Learning Python is fun!" 
# exactly 7 times.
for _ in range(7):
    print('Learning Python is fun!')

# Additional Coding Challenges
# Challenge A: Sum of Odd Numbers
# Write a program that initializes a list of numbers from 1 to 15. 
# Then, use a for loop to calculate and print the sum of only the 
# odd numbers in that list.
numbers = list(range(0, 16))
empty = 0
for num in numbers:
    if num % 2 != 0:
        empty += num

print(empty)

# Challenge B: Celsius Conversion Table
# Ask the user to enter a starting temperature in Celsius (e.g., 0).
# Then, use a for loop and the range() function to print a conversion
# table for temperatures from the user's starting point up to 10 degrees
# higher. For each Celsius temperature in this range, calculate and print
# its Fahrenheit equivalent. The formula for Celsius to Fahrenheit
# is: F = C * 9/5 + 32.
# Your output should be formatted clearly, e.g., "0°C is 32.0°F".

start_celsius = int(input('Enter a starting temperature in Celsius: '))

# Loop for 11 steps (0 to 10 degrees higher)
for offset in range(11):
    current_celsius = start_celsius + offset
    fahrenheit = (current_celsius * 9/5) + 32
    print(f'{current_celsius}°C is {fahrenheit}°F')

# Challenge C: Consonant Counter
# Ask the user to enter a sentence. Use a for loop to count how many 
# times a consonant (any letter that is not a 'a', 'e', 'i', 'o', 'u', 
# case-insensitive) appears in the sentence. Print the total count.
# (Hint: You might want to consider how to handle spaces, punctuation, 
# or numbers if they appear in the sentence – focus on just letters first.)# Challenge C: Consonant Counter
# Ask the user to enter a sentence. Use a for loop to count how many 
# times a consonant (any letter that is not a 'a', 'e', 'i', 'o', 'u', 
# case-insensitive) appears in the sentence. Print the total count.
# (Hint: You might want to consider how to handle spaces, punctuation, 
# or numbers if they appear in the sentence – focus on just letters first.)
word = input('Enter word;').lower()
vowel = ('a', 'e', 'i', 'o', 'u')
count = 0
for char in word:
    if char.isalpha():
        if char not in vowel:
            count += 1

print(f'Total consonants found: {count}')
```
