```python
# Game: Treasure Room
# You're in a room with several items. Some are useful, and some are junk. Your job is to collect only the useful items.

# üß∞ The Setup
python
Copy code
items_in_room = {
    "rusty sword": "junk",
    "gold coin": "useful",
    "old boot": "junk",
    "health potion": "useful",
    "broken arrow": "junk",
    "magic scroll": "useful"
}

inventory = []

for item, item_type in items_in_room.items():
    if item_type == "useful":
        print(f"You picked up the {item}.")
        inventory.append(item)

print("\nYour inventory contains:")
for i in inventory:
    print(f"- {i}")
```

```python
# Challenge: Monster Tracker
# You're a ranger exploring the forest, and you've spotted some creatures. Some are harmless, but some are dangerous monsters. Your task is to:

# Loop through a dictionary of creatures.

# Print a warning only for the dangerous ones.

# Add the dangerous ones to a list called threats.

# After checking all creatures, print the total number of threats and list them.

# üß™ The Data You'll Work With
# Use a dictionary like this:

# python
# Copy code
# creatures = {
#     "deer": "harmless",
#     "goblin": "dangerous",
#     "rabbit": "harmless",
#     "troll": "dangerous",
#     "butterfly": "harmless",
#     "dragon": "dangerous"
# }
# üß† What Your Program Should Do:
# Loop through the dictionary.

# For each creature:

# If it's "dangerous", print:
# ‚ö†Ô∏è Danger! A {creature} has been spotted!

# Add it to the threats list.

# After the loop:

# Print how many threats were found.

# Print each threat in the list.
creatures = {
    "deer": "harmless",
    "goblin": "dangerous",
    "rabbit": "harmless",
    "troll": "dangerous",
    "butterfly": "harmless",
    "dragon": "dangerous"
}

threats = []

for k, v in creatures.items():
    if v == 'dangerous':
        print(f'Danger! A {k} has been spotted')
        threats.append(k)

print(f"\nTotal threats detected: {len(threats)}")

for c in threats:
    print(f'{c} appears {threats.count(c)}')
```

```python
# Challenge: Potion Inventory
# You are a wizard managing your potions. You have a dictionary of potions with their quantities:

# python
# Copy code
# potions = {
#     "healing": 3,
#     "mana": 5,
#     "strength": 0,
#     "invisibility": 2,
#     "poison": 0
# }
# What your program should do:
# Loop through the potions dictionary.

# For each potion:

# If the quantity is 0, print:
# ‚ö†Ô∏è You are out of {potion} potion!

# Otherwise, print:
# You have {quantity} {potion} potion(s).

# Keep track of which potions are out of stock in a list called empty_potions.

# After the loop, print:
# Total potions out of stock: {count}
# and list the names of the empty potions one per line.

# Example output:
# scss
# Copy code
# You have 3 healing potion(s).
# You have 5 mana potion(s).
# ‚ö†Ô∏è You are out of strength potion!
# You have 2 invisibility potion(s).
# ‚ö†Ô∏è You are out of poison potion!

# Total potions out of stock: 2
# - strength
# - poison

potions = {
    "healing": 3,
    "mana": 5,
    "strength": 0,
    "invisibility": 2,
    "poison": 0
}

empty_potions = []
for k, v in potions.items():
    if v == 0:
        print(f'You are out of {k} potion!')
        empty_potions.append(k)

print(f'Total potions out of stock: {len(empty_potions)}')

for p in empty_potions:
    print(f'- {p}')
```

```python
# Challenge: Spellbook Check
# You're a wizard preparing for battle. You have a dictionary of spells and their status. Some spells are ready, others are on cooldown.

# python
# Copy code
# spells = {
#     "fireball": "ready",
#     "lightning bolt": "cooldown",
#     "heal": "ready",
#     "ice blast": "cooldown",
#     "teleport": "ready"
# }
# üß† Your Task:
# Loop through the spells dictionary.

# For each spell:

# If the spell is on "cooldown", print:
# ‚ö†Ô∏è {spell} is on cooldown!

# Otherwise, print:
# {spell} is ready to cast.

# Keep track of all spells on cooldown in a list called cooldown_spells.

# After the loop:

# Print the total number of spells on cooldown.

# List each cooldown spell on a new line with a dash (e.g., - lightning bolt).

# ‚ú® Example Output:
# pgsql
# Copy code
# fireball is ready to cast.
# ‚ö†Ô∏è lightning bolt is on cooldown!
# heal is ready to cast.
# ‚ö†Ô∏è ice blast is on cooldown!
# teleport is ready to cast.

# Total spells on cooldown: 2
# - lightning bolt
# - ice blast

spells = {
    "fireball": "ready",
    "lightning bolt": "cooldown",
    "heal": "ready",
    "ice blast": "cooldown",
    "teleport": "ready"
}

cooldown_spells = []

for k, v in spells.items():
    if v == 'cooldown':
        print(f'{k} is on cooldown')
        cooldown_spells.append(k)
    else:
        print(f'{k} is ready')

print(f'Total # of spells on cooldown {len(cooldown_spells)}')

for i in cooldown_spells:
    print(f'-{i}')
```

```python
# Challenge: Monster Battle Arena
# You're entering a battle arena. A random monster appears each round, and you must decide whether to fight, run, or use a potion.

# üß† What You'll Practice:
# Looping + conditionals ‚úÖ

# input() for player choices üßë‚Äçüíª

# random.choice() to spawn monsters üé≤

# Lists and dictionaries üß†

# Game-like flow and decision handling üí¨

# üßæ Setup:
# You start with:

# python
# Copy code
# monsters = ["goblin", "orc", "troll", "dragon", "slime"]
# potions = {
#     "healing": 2,
#     "strength": 1
# }
# üéØ What Your Program Should Do:
# Each round, randomly choose a monster from the list.

# Ask the player:

# arduino
# Copy code
# A wild {monster} appears! What do you want to do? (fight/run/use potion):
# If they fight:

# Print "You fight the {monster} bravely!"

# If they run:

# Print "You escape from the {monster}... this time."

# If they use potion:

# Show available potions and ask which one to use.

# If they enter a valid potion and have at least 1, reduce the count and print:
# "You used a {potion} potion!"

# If they don‚Äôt have that potion, print:
# "No {potion} potions left!"

# After 3 rounds, print:

# "Battle complete!"

# List remaining potions.

# üìù Hints:
# Use import random and random.choice(monsters)

# Use input() to get the player‚Äôs decision

# Use a for loop (or range(3)) to simulate 3 rounds

# Use .get(potion_name, 0) to safely check inventory

import random

monsters = ["goblin", "orc", "troll", "dragon", "slime"]
potions = {
    "healing": 2,
    "strength": 1
}

# Loop for 3 rounds of battle
for round_num in range(1, 4):
    print(f"\n--- Round {round_num} ---")
    
    # Pick a random monster
    monster = random.choice(monsters)
    
    # Ask player what to do
    user = input(f'A wild {monster} appears! What do you want to do? (fight/run/use potion): ').lower()
    
    if user == "fight":
        print(f"You fight the {monster} bravely!")
    elif user == "run":
        print(f"You escape from the {monster}... this time.")
    elif user == "use potion":
        print("Your potions:")
        for k, v in potions.items():
            print(f'- {k} ({v})')
        choice = input('Which potion do you want to use?: ').lower()
        
        if potions.get(choice, 0) > 0:
            potions[choice] -= 1
            print(f'You used a {choice} potion!')
        else:
            print(f'No {choice} potions left!')
    else:
        print("Invalid action.")

# After all rounds, print summary of potions left
print("\n--- Battle complete! ---")
print("Potions left:")
for k, v in potions.items():
    print(f'- {k}: {v}')
```
