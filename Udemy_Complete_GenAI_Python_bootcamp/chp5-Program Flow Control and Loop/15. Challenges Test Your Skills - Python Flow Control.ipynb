```python
# Challenge #1

# Write a Python script that asks the user for a number and prints a
# list of all its divisors for each number less than the given number.

user_num = int(input('Enter a #: '))

# Loop through each number from 1 up to (but not including) user_num
for num in range(1, user_num):
    # Initialize an empty list to store divisors for the current 'num'
    divisors = []

    # Inner loop to check for potential divisors from 1 up to 'num'
    for i in range(1, num + 1):
        # Check if 'i' divides 'num' evenly
        if num % i == 0:
            # If it does, add 'i' to the divisors list
            divisors.append(i)

    # After checking all potential divisors for 'num', print the result
    print(f"Divisors of {num}: {divisors}")

# Challenge #2
# Write a Python program to check if an integer x is a power of
# another integer y. Prompt the user to input both numbers.
# Input: 16, 2
# Output: 2 ** 4 = 16

# Solution from course
x = int(input("Enter a number x: "))
y = int(input(f"Enter a number y to test if x which is {x} is a power of y: "))

found = False # Flag to indicate if a power was found

# Loop through possible exponents (k) starting from 2 up to x // 2
for k in range(2, x//2): 
    if y ** k == x: # Directly checks if y to the power of k equals x
        print(f"{y} ** {k} = {x}")
        found = True # Set flag to True
        break        # Exit the loop early because we found it
# This 'else' block executes ONLY if the 'for' loop finishes naturally (i.e., no 'break' was encountered)
else: 
    print(f'{x} is not the power of {y}')
# Solution using Gemini
x = int(input('Enter a 1st #: '))
y = int(input('Enter 2nd #: '))

counter = 0  # To count the exponent
original_x = x # Store the original value of x before it gets modified

# Handle edge cases first
if x == 1:
    # 1 is y to the power of 0, for any non-zero y
    print(f'Output: {y} ** 0 = 1')
elif y == 0:
    # Cannot divide by zero
    print('Cannot divide by 0')
elif y == 1:
    # If y is 1, x can only be a power of y if x is also 1 (handled above)
    print(f'{original_x} is not a power of {y}')
else:
    # Main logic: Repeatedly divide x by y as long as it's perfectly divisible
    while x % y == 0:
        x //= y  # Integer division to reduce x
        counter += 1 # Increment the exponent counter

    # After the loop, check if x has been reduced to 1
    if x == 1:
        # If x is 1, it means original_x was a perfect power of y
        print(f'Output: {y} ** {counter} = {original_x}')
    else:
        # If x is not 1, original_x was not a perfect power of y
        print(f'{original_x} is not a power of {y}')

# Challenge #3
# Write a Python program that counts and displays the vowels of a given string, ignoring the letter case.
# Input str: Hello Everybody!
# Output: 5
sent = input('Enter a sentence:').lower()
vowels = ('a', 'e', 'i', 'o', 'u')
count = 0
for char in sent:
    if char in vowels:
        count += 1
print(F'String # of vowels {count}')

# Challenge #4
# Write a Python script that checks whether a triangle is equilateral, 
# isosceles, or scalene.
# Prompt the user to enter the lengths of the three sides.
# Triangle Types:
# Equilateral: All three sides are equal.
# Isosceles: Two sides are equal.
# Scalene: All sides are different.
# Input: Enter the lengths of the triangle sides:

# x: 6

# y: 8

# z: 12

# Output: Scalene triangle.
x = int(input('Enter a num for x:'))
y = int(input('Enter a num for y:'))
z = int(input('Enter a num for z:'))
if x + y > z and x + z > y and y + z > x:
    if x == y and y==z and z==x:
        print('Equilateral Triangle')
    elif x == y or y == z or z == x:
        print('Isosceles Triangle')
    else:
        print('Scalene Triangle')


# Challenge #5
# Write a Python program that prompts the user for multiple float numbers and
# calculates:
# The sum
# The product
# The average
# Enter 0 to finish.
# Course Solution
print("Enter some floats to calculate their sum, product and average. Input 0 to exit.")

count = 0
sum = 0.0
product = 1

while True:
	number = float(input(''))
	if number == 0:
		break

	sum += number
	product *= number
	count += 1

if count < 2:
	print("Inter at least two numbers.")
else:
	print(f'Sum, product and average are: {sum}, {sum/count}, {product}')
# Gemini Solution
# Initialize variables to store the running totals
total_sum = 0.0  # Use float for sum to match float inputs
total_product = 1.0 # Start product at 1.0 to ensure correct multiplication with floats
count_of_numbers = 0

# Get the first number from the user
# This number will also be used to control the while loop
current_num = float(input('Enter a number (or 0 to finish): '))

# Loop as long as the entered number is not 0
while current_num != 0:
    # Add the current number to the sum
    total_sum += current_num

    # Multiply the current number into the product
    total_product *= current_num

    # Increment the count of numbers entered
    count_of_numbers += 1

    # Get the next number from the user
    # This updates current_num for the next loop iteration's condition check
    current_num = float(input('Enter another number (or 0 to finish): '))

# After the loop finishes (when 0 is entered):
# Check if any numbers were entered before calculating the average
if count_of_numbers > 0:
    average = total_sum / count_of_numbers
    print("\n--- Results ---")
    print(f"The sum of the numbers is: {total_sum}")
    print(f"The product of the numbers is: {total_product}")
    print(f"The average of the numbers is: {average}")
else:
    # This block executes if the user entered 0 as their very first input
    print("\nNo numbers were entered (excluding the 0 to finish).")
    print(f"The sum is: {total_sum}")       # Will be 0.0
    print(f"The product is: {total_product}") # Will be 1.0
    # Average cannot be calculated if no numbers were entered


# Challenge #6
# Given a string, write a program that calculates the sum and average
# of all digits in the string, ignoring other characters.
# Example:
# Input: "Python31py50"
# Output: Sum: 9, Average: 2.25
# Gemini Solution
sent = input('Enter a sent:').lower() # Get input string and convert to lowercase

count = 0        # To count the number of digits found
sum = 0          # To sum the values of the digits
found_digit = False # Flag to track if any digits were encountered

# Iterate through each character in the input string
for char in sent:
    # Check if the current character is a digit
    if char.isdigit():
        found_digit = True # Set the flag to True as a digit has been found
        num = int(char)    # Convert the digit character to an integer
        sum += num         # Add the digit's value to the sum
        count += 1         # Increment the count of digits

# After the loop, check if any digits were found using the flag
if found_digit == False:
    print("No numbers were found.")
else:
    # Calculate the average (safe to divide by 'count' because 'found_digit' means 'count' > 0)
    average = sum / count
    # Print the sum and average in the specified format
    print(f'Sum: {sum}, Average: {average}')

# Course Solution
my_str = 'Python31py50'

total, count = 0, 0
for char in my_str:
    if char.isdigit():
        total += int(char)
        count += 1

print(f'Sum: {total}\nAverage: {total/count}')


# Challenge #7
# Write a Python program that displays the multiplication table 
# (from 1 to 10) for a number entered by the user.
# Input: User enters 8
# Output:

# 8 x 1 = 8

# 8 x 2 = 16

# 8 x 3 = 24

# 8 x 4 = 32

# 8 x 5 = 40

# 8 x 6 = 48

# 8 x 7 = 56

# 8 x 8 = 64

# 8 x 9 = 72

# 8 x 10 = 80

num = int(input('Enter a number:'))
for n in range(1, 11):
    product = num * n
    print(f'{num} X {n} = {product}')


# Challenge #8
# Write a Python script that displays the following pattern from 1  to n 
# where n is entered by the user.
# If the user enters 6 it will display:

# 1

# 22

# 333

# 4444

# 55555

# 666666
n = int(input('Enter a number:'))
for i in range(1, n + 1):
    s = str(i) * i
    print(f'{s}')


# Challenge #9
# Write a Python program that finds the common characters that appear 
# in two given strings.
first_string = input('Enter a sent:').lower()
second_string = input('Enter another sent:').lower()
first_set = set(first_string)
second_set = set(second_string)
result = first_set & second_set
print(result)


# Challenge #10

# Write a Python program that iterates through numbers from 1 to 50 and prints:
# "Foo" for multiples of 3
# "Bar" for multiples of 5
# "FooBar" for multiples of both 3 and 5
for num in range(1, 51):
    if num % 3 == 0 and num % 5 == 0:
        print('FooBar')
    elif num % 3 == 0:
        print('Foo')
    elif num % 5 == 0:
        print('Bar')
    else:
        print(f'{num} not a multiple of 3 or 5')


# Challenge #11
# Write a Python script that prints out the Fibonacci series up to a 
# given number n.
# Fibonacci Series: https://www.mathsisfun.com/numbers/fibonacci-sequence.html
# Example: if n is 23 it will print out 0, 1, 1, 2, 3, 5, 8, 13, 21
n = int(input('Enter a num:'))
first_num = 0
second_num = 1

if n == 0:
    print(first_num) # If n is 0, the series is just 0
else:
    # Always print the first number (0)
    print(first_num, end='') # Prints 0, no comma yet

    # Check if 'n' is large enough to include 'second_num' (1)
    if n >= 1:
        print(f', {second_num}', end='') # Print '1' with a leading comma and space

        # Calculate the third Fibonacci number (0 + 1 = 1)
        next_fib = first_num + second_num

        # Loop as long as the calculated 'next_fib' is less than or equal to 'n'
        while next_fib <= n:
            print(f', {next_fib}', end='') # Print the current 'next_fib' with a comma

            # Update first_num and second_num for the next iteration
            # This is where your correct line goes!
            first_num, second_num = second_num, next_fib

            # Now, calculate the *new* next_fib based on the updated first_num and second_num
            next_fib = first_num + second_num

    # After the loop, print a newline character to finish the line (optional but good practice)
    print()

# Challenge #12
# Write a Python script that draws the following pattern using for loops.

# Part 1: Increasing pattern (1 to 5 asterisks)
for i in range(1, 6):
    print('* ' * i)

# Part 2: Decreasing pattern (4 down to 1 asterisk)
# Starts at 4, goes down to (but not including) 0, stepping by -1
for x in range(4, 0, -1):
    print('* ' * x)
```
