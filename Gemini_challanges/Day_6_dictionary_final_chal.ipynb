```python
creatures = {
    "deer": "harmless",
    "goblin": "dangerous",
    "rabbit": "harmless",
    "troll": "dangerous",
    "butterfly": "harmless",
    "dragon": "dangerous"
}

threats = []

for k, v in creatures.items():
    if v == 'dangerous':
        print(f'Danger! A {k} has been spotted')
        threats.append(k)

print(f"\nTotal threats detected: {len(threats)}")

for c in threats:
    print(f'{c} appears {threats.count(c)}')
```
