```python
# 1. Hello World
# Print "Hello, World!" to the screen.
print('Hello, World!')  # Output: Hello, World!
# 2. Print Your Name
# Print your full name on the screen.
name = input("Enter your full name: ")  # Prompt user for their full name
print("Your name is:", name)  # Output: Your name is: [name]
# 3. Personalized Greeting
# Ask the user for their name and greet them with:
# "Hello, [name]!"
user_name = input("What is your name? ")  # Prompt user for their name
print(f"Hello, {user_name}!")  # Output: Hello, [name]
# 4. Favorite Color
# Ask the user for their favorite color and print a message like:
# "Wow! [color] is a great choice!"
color = input("What is your favorite color? ")  # Prompt user for their favorite color
print(f"Wow! {color} is a great choice!")  # Output: Wow! [color] is a great choice!
# 5. Add Two Numbers
# Ask the user to enter two numbers and print their sum.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
sum_result = num1 + num2  # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 6. Subtract Two Numbers
# Ask the user to enter two numbers and print the difference.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
sum_result = num1 - num2  # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 7. Multiply Two Numbers
# Ask the user to enter two numbers and print the product.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
sum_result = num1 * num2  # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 8. Divide Two Numbers
# Ask the user to enter two numbers and print the result of the division.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
sum_result = num1 / num2  # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 9. Modulo Operation
# Ask the user for two numbers and print the remainder of the division (use % operator).
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
sum_result = num1 % num2  # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 10. Square a Number
# Ask the user for a number and print its square.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
sum_result = num1 ** num2  # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 11. Area of a Rectangle
# Ask the user for the width and height of a rectangle and calculate the area.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
sum_result = 0.5 * num1 * num2  # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 12. Convert Minutes to Seconds
# Ask the user for a number of minutes and convert it to seconds.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
sum_result = num1 * 60.0 # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 13. Age Calculator
# Ask for the user’s birth year and print their age assuming the current year is 2025.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
sum_result =   2025 - num1 # Calculate the sum of the two numbers
print(f'Your total is: {sum_result}') # Output: Your total is: [sum_result]
# 14. Temperature Converter
# Ask the user for a temperature in Celsius and convert it to Fahrenheit using the formula:
# F = (C × 9/5) + 32
celsius = float(input("Enter temperature in Celsius: "))  # Prompt user for temperature in Celsius
fahrenheit = (celsius * 9/5) + 32  # Convert Celsius to Fahrenheit
print(f"Temperature in Fahrenheit: {fahrenheit}")  # Output: Temperature in Fahrenheit: [fahrenheit]
# 15. Simple Interest Calculator # Not using split function
# Ask for principal, rate, and time. Calculate simple interest:
# SI = (P × R × T) / 100
p = float(input('Enter principle amount:'))  # Prompt user for principal amount using float
r = float(input('Enter rate of interest:'))  # Prompt user for rate of interest using float
t = float(input('Enter time in years:'))  # Prompt user for time in years using float
simple_interest = (p * r * t) / 100  # Calculate simple interest
print(f'THe simple interest is: {simple_interest}')  # Output: The simple interest is: [simple_interest]
# 16. Perimeter of a Square
# Ask the user for the side length of a square and calculate its perimeter.
length = float(input("Enter the side length of the square: "))  # Prompt user for side length
perimeter = 4 * length  # Calculate the perimeter of the square
print(f"The perimeter of the square is: {perimeter}")  # Output: The perimeter of the square is: [perimeter]
# 17. Average of Three Numbers
# Ask the user for three numbers and print their average.
p = float(input('Enter principle amount:'))  # Prompt user for principal amount using float
r = float(input('Enter rate of interest:'))  # Prompt user for rate of interest using float
t = float(input('Enter time in years:'))  # Prompt user for time in years using float
average = (p * r * t) / 3  # Calculate average of three numbers
print(f'THe average is: {average}')  # Output: The average is:
# 18. Swap Two Numbers
# Ask for two numbers and swap their values. Print the values before and after swapping.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
print(f"Before swapping: num1 = {num1}, num2 = {num2}")  # Output before swapping
num1, num2 = num2, num1  # Swap the values of num1 and num2
print(f"After swapping: num1 = {num1}, num2 = {num2}")  # Output after swapping
# 19. Repeat a Word
# Ask the user for a word and a number. Print the word that many times.
word = input("Enter a word: ")  # Prompt user for a word
number = int(input("How many times do you want to repeat it? "))  # Prompt user for a number
print(word * number)  # Output: [word repeated number times]
# 20. Simple Calculator Menu -DIFFICULT if statements have not been covered yet
# Ask the user for two numbers and let them choose an operation: add, subtract, multiply, divide. Perform and print the result.
num1 = float(input("Enter the first number: "))  # Prompt user for the first number
num2 = float(input("Enter the second number: "))  # Prompt user for the second number
operation = input("Choose an operation (add, subtract, multiply, divide): ").strip().lower()  # Prompt user for an operation
if operation == "add":
    result = num1 + num2  # Perform addition
elif operation == "subtract":
    result = num1 - num2    # Perform subtraction
elif operation == "multiply":   
    result = num1 * num2  # Perform multiplication
elif operation == "divide":
    if num2 != 0:
        result = num1 / num2  # Perform division
    else:
        result = "Error: Division by zero is not allowed."  # Handle division by zero
else:
    result = "Error: Invalid operation."    # Handle invalid operation
print(f"The result is: {result}")  # Output: The result is: [result]
# Code from chatgpt

print("Welcome to the Simple Calculator!")
print("Choose an operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

# Get user input
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
choice = input("Enter your choice (1/2/3/4): ")

# Perform selected operation
if choice == '1':
    result = num1 + num2
    print(f"The sum is: {result}")
elif choice == '2':
    result = num1 - num2
    print(f"The difference is: {result}")
elif choice == '3':
    result = num1 * num2
    print(f"The product is: {result}")
elif choice == '4':
    if num2 == 0:
        print("Error: Cannot divide by zero.")
    else:
        result = num1 / num2
        print(f"The quotient is: {result}")
else:
    print("Invalid choice. Please select 1, 2, 3, or 4.")
```
